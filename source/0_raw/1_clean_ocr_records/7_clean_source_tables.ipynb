# Streamlined Geocoding Data Extraction Notebook

This notebook processes geocoding data and extracts additional information from complex address strings. It has been streamlined to:

1. **Preserve existing data** - The extraction functions check for existing values and only add new data without removing/replacing existing entries in the "Exit_Number_2" and "Secondary_Road" columns
2. **Remove redundancies** - Eliminated duplicate code blocks that perform similar tasks
3. **Improve extraction patterns** - Enhanced regex patterns to better handle different address formats

The notebook extracts the following information from complex address strings:
- Secondary_Road (from addresses with slashes like "I-80 Exit 149 EB / I-15 WB")
- Exit_Number_2 (additional exit numbers after slashes)
- Tertiary_Road (third road in complex intersections)
- Exit_Number_3 (exit numbers for tertiary roads)

```python
from pathlib import Path

def find_project_root(start=None, markers=("run_all.py", "requirements.txt")):
    p = Path(start or Path.cwd()).resolve()
    for candidate in (p, *p.parents):
        if any((candidate / m).exists() for m in markers):
            return candidate
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

import pandas as pd

df1 = pd.read_csv(DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/6_extract_exit_numbers.csv")
df1
# Drop the address column
df1 = df1.drop(columns=['address'])

# Display the updated DataFrame
df1
```

```python
import re

def clean_chain_name(name):
    """Normalize chain names for the scraped dataset (same logic as OCR pipeline step 5)."""
    if pd.isna(name):
        return name

    cleaned = str(name).strip()
    lowered = cleaned.lower()

    if lowered.startswith("bp-"):
        return "bp"
    if lowered.startswith("shell_"):
        return "shell"
    if "circle_k" in lowered and "_" in lowered:
        return "circle k"

    direct_map = {
        "66gas": "66 gas", "76": "76", "76_gas": "76", "76gas": "76",
        "gulf_oil": "gulf oil", "flying-j": "flying j", "flying-j-": "flying j",
        "qt": "quicktrip", "quiktrip": "quicktrip", "quicktrip": "quicktrip",
        "kum_go": "kum & go", "kum go": "kum & go", "kumgo": "kum & go",
        "phillips_66": "phillips 66", "phillips66": "phillips 66", "phillips 66": "phillips 66",
        "exxonmobil": "exxon", "exxon_mobil": "exxon", "exxon": "exxon",
        "mobil": "mobil", "mobil_gas": "mobil",
        "chevron": "chevron", "chevron_gas": "chevron",
        "texaco": "texaco", "texaco_gas": "texaco",
        "arco": "arco", "arco_gas": "arco",
        "valero": "valero", "valero_gas": "valero",
        "sunoco": "sunoco", "sunoco_gas": "sunoco",
        "citgo": "citgo", "citgo_gas": "citgo",
        "marathon": "marathon", "marathon_gas": "marathon",
        "speedway": "speedway", "speedway_gas": "speedway",
        "wawa": "wawa", "wawa_gas": "wawa",
        "sheetz": "sheetz", "sheetz_gas": "sheetz",
        "casey\'s": "casey\'s", "caseys": "casey\'s", "casey_s": "casey\'s",
        "holiday": "holiday", "holiday_gas": "holiday",
        "pilot": "pilot", "pilot_gas": "pilot", "pilot travel center": "pilot",
        "loves": "love\'s", "love\'s": "love\'s", "loves_gas": "love\'s",
        "ta": "ta", "ta_gas": "ta", "travelcenters": "ta",
        "petro": "petro", "petro_gas": "petro",
        "sinclair": "sinclair", "sinclair_gas": "sinclair",
        "conoco": "conoco", "conoco_gas": "conoco", "conocophillips": "conoco",
    }

    if lowered in direct_map:
        return direct_map[lowered]

    cleaned = re.sub(r"_+\d*$", "", cleaned)
    cleaned = re.sub(r"^-+|-+$", "", cleaned)
    cleaned = re.sub(r"_+", " ", cleaned)
    cleaned = re.sub(r"-+", " ", cleaned)
    cleaned = re.sub(r"\s+", " ", cleaned).strip().lower()
    return cleaned

# Normalize chain names in scraped data
if 'Chain' in df2.columns:
    before = df2['Chain'].nunique(dropna=True)
    df2['Chain'] = df2['Chain'].apply(clean_chain_name)
    after = df2['Chain'].nunique(dropna=True)
    print(f"Normalized scraped Chain: unique values {before} -> {after}")

#save df2 as ../../../data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/7_clean_source_tables_scraped.csv
df2.to_csv(DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/7_clean_source_tables_scraped.csv", index=False)
```
