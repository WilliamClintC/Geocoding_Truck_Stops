# 4_extract_road_info

Extract `Main_Road` and `Secondary_Road` from cleaned address text.

## Inputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/3_clean_exit_addresses.csv`

## Outputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/4_extract_road_info.csv`

## Transformations
1. Parse road segments from `address_standardized_OFF_parenthesis`.
2. Support highways, route numbers, and named streets.
3. Store extracted road pieces in `Main_Road` and `Secondary_Road`.

```python
from pathlib import Path

def find_project_root(start=None, markers=("run_all.py", "requirements.txt")):
    p = Path(start or Path.cwd()).resolve()
    for candidate in (p, *p.parents):
        if any((candidate / m).exists() for m in markers):
            return candidate
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

from pathlib import Path
import re

import pandas as pd

INPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/3_clean_exit_addresses.csv"
OUTPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/4_extract_road_info.csv"


def extract_single_road(text):
    if pd.isna(text) or not str(text).strip():
        return ""

    road_text = str(text).strip()

    highway_patterns = [
        r"\b(Hwy\s+\d+(?:\s+[NSEW])?)",
        r"\b(US\s+\d+)",
        r"\b(CA\s+\d+(?:-\d+)?)",
        r"\b(I-\d+(?:-\d+)?)",
        r"\b(SR\s+\d+)",
        r"\b(UT\s+\d+)",
        r"\b(NV\s+\d+)",
        r"\b(AZ\s+\d+)",
    ]
    for pattern in highway_patterns:
        match = re.search(pattern, road_text, flags=re.IGNORECASE)
        if match:
            return match.group(1).strip()

    abbrev_match = re.search(r"\b((?:Rd|St|Ave|Blvd|Dr|Ln|Way|Pkwy)\s+\w+)\b", road_text, flags=re.IGNORECASE)
    if abbrev_match:
        return abbrev_match.group(1).strip()

    numbered_match = re.search(
        r"\b(\d+(?:st|nd|rd|th)\s+(?:St|Street|Ave|Avenue|Rd|Road|Blvd|Boulevard))\b",
        road_text,
        flags=re.IGNORECASE,
    )
    if numbered_match:
        return numbered_match.group(1).strip()

    named_match = re.search(
        r"\b([A-Za-z]+(?:\s+[A-Za-z]+)*)\s+(St|Street|Ave|Avenue|Rd|Road|Blvd|Boulevard|Dr|Drive|Ln|Lane|Way|Pkwy|Parkway)\b",
        road_text,
        flags=re.IGNORECASE,
    )
    if named_match:
        return f"{named_match.group(1).strip()} {named_match.group(2).strip()}"

    words = re.findall(r"\b\w+\b", road_text)
    stop_words = {"exit", "jct", "junction", "at", "near", "and", "the", "on", "off", "to", "from"}
    filtered = [w for w in words if w.lower() not in stop_words]
    return " ".join(filtered[:2]) if filtered else ""


def extract_road_info(address):
    if pd.isna(address) or not str(address).strip():
        return "", ""

    text = str(address).strip()

    hyphen_match = re.search(r"\b((?:CA|US|I-|Hwy|SR)\s*\d+-\d+)\b", text, flags=re.IGNORECASE)
    if hyphen_match and "&" not in text:
        return hyphen_match.group(1).strip(), ""

    if "&" in text:
        left, right = text.split("&", 1)
        return extract_single_road(left), extract_single_road(right)

    return extract_single_road(text), ""


df = pd.read_csv(INPUT_PATH)
print(f"Loaded {len(df):,} rows from {INPUT_PATH}")

if "address_standardized_OFF_parenthesis" not in df.columns:
    raise ValueError("Missing required column: address_standardized_OFF_parenthesis")

roads = df["address_standardized_OFF_parenthesis"].apply(extract_road_info)
df["Main_Road"] = roads.apply(lambda x: x[0])
df["Secondary_Road"] = roads.apply(lambda x: x[1])

Path(OUTPUT_PATH).parent.mkdir(parents=True, exist_ok=True)
df.to_csv(OUTPUT_PATH, index=False)
print(f"Saved road extraction output to {OUTPUT_PATH}")

df[["address_standardized_OFF_parenthesis", "Main_Road", "Secondary_Road"]].head()
```
