# 5_normalize_chains

Normalize chain names into consistent values.

## Inputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/4_extract_road_info.csv`

## Outputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/5_normalize_chains.csv`

## Transformations
1. Detect available chain column (`chain`, `Chain`, `OCR_chain`).
2. Apply canonical naming rules for major brands.
3. Save normalized chain text in place.

```python
from pathlib import Path

def find_project_root(start=None, markers=("run_all.py", "requirements.txt")):
    p = Path(start or Path.cwd()).resolve()
    for candidate in (p, *p.parents):
        if any((candidate / m).exists() for m in markers):
            return candidate
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

from pathlib import Path
import re

import pandas as pd

INPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/4_extract_road_info.csv"
OUTPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/5_normalize_chains.csv"


def pick_column(df, candidates):
    for col in candidates:
        if col in df.columns:
            return col
    return None


def clean_chain_name(name):
    if pd.isna(name):
        return name

    cleaned = str(name).strip()
    lowered = cleaned.lower()

    # Prefix rules
    if lowered.startswith("bp-"):
        return "bp"
    if lowered.startswith("shell_"):
        return "shell"
    if "circle_k" in lowered and "_" in lowered:
        return "circle k"

    # Direct brand mappings (comprehensive list)
    direct_map = {
        "66gas": "66 gas",
        "76": "76",
        "76_gas": "76",
        "76gas": "76",
        "gulf_oil": "gulf oil",
        "flying-j": "flying j",
        "flying-j-": "flying j",
        "qt": "quicktrip",
        "quiktrip": "quicktrip",
        "quicktrip": "quicktrip",
        "kum_go": "kum & go",
        "kum go": "kum & go",
        "kumgo": "kum & go",
        "phillips_66": "phillips 66",
        "phillips66": "phillips 66",
        "phillips 66": "phillips 66",
        "exxonmobil": "exxon",
        "exxon_mobil": "exxon",
        "exxon": "exxon",
        "mobil": "mobil",
        "mobil_gas": "mobil",
        "chevron": "chevron",
        "chevron_gas": "chevron",
        "texaco": "texaco",
        "texaco_gas": "texaco",
        "arco": "arco",
        "arco_gas": "arco",
        "valero": "valero",
        "valero_gas": "valero",
        "sunoco": "sunoco",
        "sunoco_gas": "sunoco",
        "citgo": "citgo",
        "citgo_gas": "citgo",
        "marathon": "marathon",
        "marathon_gas": "marathon",
        "speedway": "speedway",
        "speedway_gas": "speedway",
        "wawa": "wawa",
        "wawa_gas": "wawa",
        "sheetz": "sheetz",
        "sheetz_gas": "sheetz",
        "casey\'s": "casey\'s",
        "caseys": "casey\'s",
        "casey_s": "casey\'s",
        "holiday": "holiday",
        "holiday_gas": "holiday",
        "pilot": "pilot",
        "pilot_gas": "pilot",
        "pilot travel center": "pilot",
        "loves": "love\'s",
        "love\'s": "love\'s",
        "loves_gas": "love\'s",
        "ta": "ta",
        "ta_gas": "ta",
        "travelcenters": "ta",
        "petro": "petro",
        "petro_gas": "petro",
        "sinclair": "sinclair",
        "sinclair_gas": "sinclair",
        "conoco": "conoco",
        "conoco_gas": "conoco",
        "conocophillips": "conoco",
    }

    if lowered in direct_map:
        return direct_map[lowered]

    # Generic cleanup
    cleaned = re.sub(r"_+\d*$", "", cleaned)
    cleaned = re.sub(r"^-+|-+$", "", cleaned)
    cleaned = re.sub(r"_+", " ", cleaned)
    cleaned = re.sub(r"-+", " ", cleaned)
    cleaned = re.sub(r"\s+", " ", cleaned).strip().lower()

    return cleaned

df = pd.read_csv(INPUT_PATH)
print(f"Loaded {len(df):,} rows from {INPUT_PATH}")

chain_col = pick_column(df, ["chain", "Chain", "OCR_chain", "ocr_chain"])
if chain_col:
    before_unique = df[chain_col].nunique(dropna=True)
    df[chain_col] = df[chain_col].apply(clean_chain_name)
    after_unique = df[chain_col].nunique(dropna=True)
    print(f"Normalized {chain_col}: unique values {before_unique} -> {after_unique}")
else:
    print("No chain column found; notebook will pass through data unchanged.")

Path(OUTPUT_PATH).parent.mkdir(parents=True, exist_ok=True)
df.to_csv(OUTPUT_PATH, index=False)
print(f"Saved chain-normalized file to {OUTPUT_PATH}")

df.head()
```
