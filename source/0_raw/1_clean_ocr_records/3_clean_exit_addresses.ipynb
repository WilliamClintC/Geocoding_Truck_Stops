# 3_clean_exit_addresses

Apply focused cleanup to `Exit`-type address strings.

## Inputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/2_identify_proper_addresses.csv`

## Outputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/3_clean_exit_addresses.csv`

## Transformations
1. Clean OCR artifacts in OFF-parenthesis address text.
2. Normalize punctuation and spacing for `Exit` records.
3. Keep non-Exit rows unchanged.

```python
from pathlib import Path

def find_project_root(start=None, markers=("run_all.py", "requirements.txt")):
    p = Path(start or Path.cwd()).resolve()
    for candidate in (p, *p.parents):
        if any((candidate / m).exists() for m in markers):
            return candidate
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

from pathlib import Path

import pandas as pd

INPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/2_identify_proper_addresses.csv"
OUTPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/3_clean_exit_addresses.csv"


df = pd.read_csv(INPUT_PATH)
print(f"Loaded {len(df):,} rows from {INPUT_PATH}")

if "Address_Type" not in df.columns:
    df["Address_Type"] = "empty"

if "address_standardized_OFF_parenthesis" not in df.columns:
    raise ValueError("Missing required column: address_standardized_OFF_parenthesis")

exit_mask = df["Address_Type"].eq("Exit")

cleaned = df.loc[exit_mask, "address_standardized_OFF_parenthesis"].astype(str)
cleaned = cleaned.str.replace(")", "", regex=False)
cleaned = cleaned.str.replace("<U+00D8>", "", regex=False)
cleaned = cleaned.str.replace(r"\s+", " ", regex=True)
cleaned = cleaned.str.replace(r"\s+,", ",", regex=True)
cleaned = cleaned.str.strip(" ,")

df.loc[exit_mask, "address_standardized_OFF_parenthesis"] = cleaned

Path(OUTPUT_PATH).parent.mkdir(parents=True, exist_ok=True)
df.to_csv(OUTPUT_PATH, index=False)
print(f"Saved cleaned exit addresses to {OUTPUT_PATH}")

df[["address_standardized_OFF_parenthesis", "Address_Type"]].head()
```
