# 1_standardize_addresses

Standardize raw address text and remove non-analytic columns.

## Inputs
- `data/0_raw/unbalanced_panel.csv`

## Outputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/1_standardized_addresses.csv`

## Transformations
1. Normalize interstate patterns (`I 80`, `I - 80`, `1-80` -> `I-80`) in all string columns.
2. Drop technical columns that are not needed downstream.
3. Save cleaned data for the next stage.

```python
from pathlib import Path

def find_project_root(start=None, markers=("run_all.py", "requirements.txt")):
    p = Path(start or Path.cwd()).resolve()
    for candidate in (p, *p.parents):
        if any((candidate / m).exists() for m in markers):
            return candidate
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

from pathlib import Path
import re

import pandas as pd

INPUT_PATH = DATA_DIR / "0_raw/unbalanced_panel.csv"
OUTPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/1_standardized_addresses.csv"


def standardize_interstate_format(text):
    """Normalize common interstate OCR variants to I-<number>."""
    if pd.isna(text) or not isinstance(text, str):
        return text

    patterns = [
        r"\bI\s*-\s*(\d+)\b",  # I-80, I - 80, I- 80
        r"\bI\s+(\d+)\b",       # I 80
        r"\b1\s*-\s*(\d+)\b",  # 1-80, 1 - 80
    ]

    cleaned = text
    for pattern in patterns:
        cleaned = re.sub(pattern, lambda m: f"I-{m.group(1)}", cleaned, flags=re.IGNORECASE)
    return cleaned


df = pd.read_csv(INPUT_PATH)
print(f"Loaded {len(df):,} rows from {INPUT_PATH}")

for column in df.columns:
    if pd.api.types.is_string_dtype(df[column]):
        df[column] = df[column].apply(standardize_interstate_format)

columns_to_drop = [
    "Unnamed: 0",
    "filename",
    "record_num",
    "parking",
    "gray_parking",
    "identifier",
    "panel",
    "p_identifier",
    "identifier2",
    "identifier3",
]

existing_to_drop = [c for c in columns_to_drop if c in df.columns]
if existing_to_drop:
    df = df.drop(columns=existing_to_drop)
    print(f"Dropped columns: {existing_to_drop}")

Path(OUTPUT_PATH).parent.mkdir(parents=True, exist_ok=True)
df.to_csv(OUTPUT_PATH, index=False)
print(f"Saved standardized file to {OUTPUT_PATH}")

df.head()
```
