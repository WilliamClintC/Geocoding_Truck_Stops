# 6_extract_exit_numbers

Extract exit numbers from address text and flag unclear OCR patterns.

## Inputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/5_normalize_chains.csv`

## Outputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/6_extract_exit_numbers.csv`

## Transformations
1. Extract `Exit_Number` from address and fallback label text.
2. Detect unclear OCR strings for `Exit` rows.
3. Update `Flagged` and `Flag_Reason` where quality issues are detected.

```python
from pathlib import Path

def find_project_root(start=None, markers=("run_all.py", "requirements.txt")):
    p = Path(start or Path.cwd()).resolve()
    for candidate in (p, *p.parents):
        if any((candidate / m).exists() for m in markers):
            return candidate
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

from pathlib import Path
import re

import pandas as pd

INPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/5_normalize_chains.csv"
OUTPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/6_extract_exit_numbers.csv"


def extract_exit_number(address_text, label_text=None):
    def _extract(text):
        if pd.isna(text) or str(text).strip() == "":
            return None

        value = str(text)
        patterns = [
            r"Exit\s+(\d+[A-Za-z]?)",
            r"\bEx\s+(\d+[A-Za-z]?)",
            r"<U\+[0-9A-Fa-f]+>xit\s+(\d+[A-Za-z]?)",
            r"[\u00C9\u00C8\u00CA\u00CB\u00E9\u00E8\u00EA\u00EB]xit\s+(\d+[A-Za-z]?)",
            r"(?:US\s+\d+|I-\d+|SR\s+\d+|CA\s+\d+|State\s+Route\s+\d+)\s+X\s+(\d+[A-Za-z]?)",
            r"(?:Highway|Hwy|Route|Rt)\s+\d+\s+X\s+(\d+[A-Za-z]?)",
            r"(?:\w+\s+)?(?:Tpke|Turnpike|Pkwy|Parkway|Expwy|Expressway|Fwy|Freeway)\s+X\s+(\d+[A-Za-z]?)",
            r"(?:\d{1,3}(?:-\d+)?)\s+X\s+(\d+[A-Za-z]?)",
        ]

        for pattern in patterns:
            match = re.search(pattern, value, flags=re.IGNORECASE)
            if match:
                return match.group(1)
        return None

    result = _extract(address_text)
    if result is not None:
        return result
    return _extract(label_text)


def is_unclear_ocr_address(address_text, address_type):
    if pd.isna(address_text) or str(address_text).strip() == "":
        return False
    if address_type != "Exit":
        return False

    text = str(address_text).strip()
    unclear_patterns = [
        r"^\d{4,}",
        r".+,.+,.+",
        r"[A-Za-z][0-9]+-[0-9]+-[0-9]+",
        r"D[A-Z][a-z]+\s+[A-Z][a-z]+\s+City",
        r"[A-Z][a-z]+\sJ\s[A-Z][a-z]+",
    ]
    for pattern in unclear_patterns:
        if re.search(pattern, text):
            return True

    problematic_phrases = ["81191-15-80", "DSalt Lake City", "nemucca,", "Flying I-", "Eagle's I-"]
    return any(phrase in text for phrase in problematic_phrases)


def pick_column(df, candidates):
    for col in candidates:
        if col in df.columns:
            return col
    return None


df = pd.read_csv(INPUT_PATH)
print(f"Loaded {len(df):,} rows from {INPUT_PATH}")

address_col = pick_column(df, ["address_standardized_OFF_parenthesis", "address", "OCR_address"])
label_col = pick_column(df, ["label", "OCR_label"])

if not address_col:
    raise ValueError("No address-like column found for exit extraction.")

if "Address_Type" not in df.columns:
    df["Address_Type"] = "empty"
if "Flagged" not in df.columns:
    df["Flagged"] = False
if "Flag_Reason" not in df.columns:
    df["Flag_Reason"] = ""

# Match reference output fields by storing extraction source-specific values.
df["Exit_From_Address"] = df[address_col].apply(lambda x: extract_exit_number(x, None))
if label_col:
    df["Exit_From_Label"] = df[label_col].apply(lambda x: extract_exit_number(x, None))
else:
    df["Exit_From_Label"] = pd.NA

# Prefer address extraction; fallback to label extraction.
df["Exit_Number"] = df["Exit_From_Address"].where(df["Exit_From_Address"].notna(), df["Exit_From_Label"])

df["Is_Unclear_OCR"] = df.apply(
    lambda row: is_unclear_ocr_address(row[address_col], row["Address_Type"]),
    axis=1,
)

unclear_mask = df["Is_Unclear_OCR"]
df.loc[unclear_mask, "Flagged"] = True
df.loc[unclear_mask, "Flag_Reason"] = "unclear address_standardized_OFF_parenthesis"

# Align exported schema to organize_reference_code/4_6.csv order.
reference_order = [
    "clean_line1",
    "clean_line2",
    "line3",
    "city",
    "zip_code",
    "label",
    "phone",
    "address",
    "year",
    "major_city",
    "state",
    "chain",
    "address_standardized_ON_parenthesis",
    "address_standardized_OFF_parenthesis",
    "Address_Type",
    "Exit_Number",
    "Exit_From_Address",
    "Exit_From_Label",
    "Flagged",
    "Flag_Reason",
    "Is_Unclear_OCR",
    "Main_Road",
    "Secondary_Road",
]

for col in reference_order:
    if col not in df.columns:
        df[col] = pd.NA

df_export = df[reference_order]

Path(OUTPUT_PATH).parent.mkdir(parents=True, exist_ok=True)
df_export.to_csv(OUTPUT_PATH, index=False)

success_count = df_export["Exit_Number"].notna().sum()
print(f"Exit numbers extracted: {success_count:,} / {len(df_export):,}")
print(f"Unclear OCR flags raised: {unclear_mask.sum():,}")
print(f"Saved final output to {OUTPUT_PATH}")

df_export.head()
```
