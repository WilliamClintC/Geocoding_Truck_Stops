# 2_identify_proper_addresses

Classify addresses into `Exit` vs `Proper` and clean extracted address text.

## Inputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/1_standardized_addresses.csv`

## Outputs
- `data/1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/2_identify_proper_addresses.csv`

## Transformations
1. Split address text into OFF/ON-parenthesis components.
2. Initialize `Address_Type`, flag likely `Exit` entries.
3. Remove ZIP code remnants and leading dashes from OFF text.
4. Mark numeric-leading entries as `Proper`.

```python
from pathlib import Path

def find_project_root(start=None, markers=("run_all.py", "requirements.txt")):
    p = Path(start or Path.cwd()).resolve()
    for candidate in (p, *p.parents):
        if any((candidate / m).exists() for m in markers):
            return candidate
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

from pathlib import Path
import re

import pandas as pd

INPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/1_standardized_addresses.csv"
OUTPUT_PATH = DATA_DIR / "1_derived/1_clean_ocr_records/1_clean_ocr_pipeline_outputs/2_identify_proper_addresses.csv"


def pick_column(df, candidates):
    for col in candidates:
        if col in df.columns:
            return col
    return None


def remove_zipcode_from_address(address_text, zip_code):
    if pd.isna(address_text):
        return address_text
    if pd.isna(zip_code):
        return address_text

    text = str(address_text)
    zip_str = str(zip_code).strip()
    if not zip_str or zip_str.lower() == "nan":
        return text

    zip_forms = {zip_str}
    if zip_str.isdigit() and len(zip_str) == 4:
        zip_forms.add("0" + zip_str)
    if zip_str.isdigit() and len(zip_str) == 5 and zip_str.startswith("0"):
        zip_forms.add(zip_str[1:])

    cleaned = text
    for z in sorted(zip_forms, key=len, reverse=True):
        cleaned = cleaned.replace(z, "")

    cleaned = " ".join(cleaned.split())
    cleaned = cleaned.replace(" ,", ",").replace(", ,", ",")
    return cleaned.strip(" ,")


df = pd.read_csv(INPUT_PATH)
print(f"Loaded {len(df):,} rows from {INPUT_PATH}")

address_col = pick_column(df, ["address", "OCR_address"])
zip_col = pick_column(df, ["zip_code", "OCR_zip_code"])

if not address_col:
    raise ValueError("Could not find an address column. Expected one of: address, OCR_address")

# Split around first parenthesis.
df["address_standardized_ON_parenthesis"] = df[address_col].astype(str).str.extract(r"(\(.*)", expand=False)
df["address_standardized_OFF_parenthesis"] = df[address_col].astype(str).str.extract(r"(.*?)(?=\(|$)", expand=False).str.strip()

if "Address_Type" not in df.columns:
    df["Address_Type"] = "empty"
else:
    df["Address_Type"] = df["Address_Type"].fillna("empty")

exit_pattern = r"([E\u00c9e\u00e9]?xit|Ex\s+\d+|Ex$|\bEx\b|\bX\s+\d+)"
exit_mask = df["address_standardized_OFF_parenthesis"].str.contains(exit_pattern, na=False, regex=True)
df.loc[exit_mask, "Address_Type"] = "Exit"

if zip_col:
    df["address_standardized_OFF_parenthesis"] = df.apply(
        lambda row: remove_zipcode_from_address(row["address_standardized_OFF_parenthesis"], row[zip_col]),
        axis=1,
    )

df["address_standardized_OFF_parenthesis"] = (
    df["address_standardized_OFF_parenthesis"]
    .astype(str)
    .str.replace(r"^-\s*", "", regex=True)
    .str.replace(r"^City,\s*", "", regex=True)
    .str.replace(")", "", regex=False)
    .str.replace("<U+00D8>", "", regex=False)
    .str.strip()
)

proper_mask = (
    (df["Address_Type"] == "empty")
    & df["address_standardized_OFF_parenthesis"].str.match(r"^\d+\s", na=False)
)
df.loc[proper_mask, "Address_Type"] = "Proper"

Path(OUTPUT_PATH).parent.mkdir(parents=True, exist_ok=True)
df.to_csv(OUTPUT_PATH, index=False)
print(f"Saved classified file to {OUTPUT_PATH}")

df[["address_standardized_OFF_parenthesis", "address_standardized_ON_parenthesis", "Address_Type"]].head()
```
