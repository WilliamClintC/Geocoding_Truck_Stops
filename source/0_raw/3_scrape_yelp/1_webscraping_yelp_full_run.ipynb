```python
import re
import random
import pandas as pd

# Function to standardize phone number to international format
def standardize_phone(phone_number):
    if pd.isna(phone_number):
        return None
    
    # Remove all non-digit characters
    digits = re.sub(r'\D', '', str(phone_number))
    
    # Make sure we have at least 10 digits
    if len(digits) < 10:
        return None
    
    # If the number starts with a 1 and has 11 digits, keep it as is
    # Otherwise, if it's 10 digits, add a 1 at the beginning
    if len(digits) == 11 and digits[0] == '1':
        return f"+{digits}"
    elif len(digits) == 10:
        return f"+1{digits}"
    elif len(digits) > 11:  # Too many digits, just use the last 10
        return f"+1{digits[-10:]}"
    else:
        return None

# Extract all unique phone numbers from the DataFrame
unique_phones = df['phone'].dropna().unique()
print(f"Total unique phone numbers before standardization: {len(unique_phones)}")

# Create a dictionary to map original phone numbers to standardized versions
phone_mapping = {}
standardized_phones = []

for phone in unique_phones:
    std_phone = standardize_phone(phone)
    if std_phone is not None:
        phone_mapping[phone] = std_phone
        standardized_phones.append(std_phone)

# Get unique standardized phone numbers
unique_standardized_phones = list(set(standardized_phones))
print(f"Total unique standardized phone numbers: {len(unique_standardized_phones)}")

# Show sample of original to standardized mapping
sample_mapping = {k: phone_mapping[k] for k in list(phone_mapping.keys())[:10]}
print("\nSample of original to standardized phone numbers:")
for orig, std in sample_mapping.items():
    print(f"{orig} -> {std}")
```

```python
import requests
import os
import time
from dotenv import load_dotenv
import json
import pandas as pd
from tqdm.auto import tqdm
import glob
from pathlib import Path

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'data').exists() and (candidate / 'source').exists() and (candidate / 'temp').exists():
            return candidate
    return start

# Load API key from .env file
PROJECT_ROOT = globals().get('PROJECT_ROOT', find_project_root(Path.cwd()))
env_file = PROJECT_ROOT / '.env'
if env_file.exists():
    load_dotenv(env_file)
else:
    load_dotenv()
api_key = os.getenv('YELP_API_KEY')

# Set up the API endpoint and headers
endpoint = 'https://api.yelp.com/v3/businesses/search/phone'
headers = {
    'Authorization': f'Bearer {api_key}'
}

def fetch_yelp_business_by_phone(phone, retry_count=0, max_retries=2):
    """
    Fetch business information from Yelp API using phone number
    
    Args:
        phone (str): Phone number in international format (+1XXXXXXXXXX)
        retry_count (int): Current retry attempt
        max_retries (int): Maximum number of retries for rate limiting
    
    Returns:
        dict: Dictionary containing business information, None if not found, or 'RATE_LIMITED' if rate limited
    """
    try:
        # Make the API request
        response = requests.get(
            endpoint,
            headers=headers,
            params={'phone': phone},
            timeout=10
        )
        
        # Check if the request was successful
        if response.status_code == 200:
            data = response.json()
            return data
        elif response.status_code == 429:
            if retry_count < max_retries:
                print(f"WARNING: Rate limited! Waiting 60 seconds... (Retry {retry_count + 1}/{max_retries})")
                time.sleep(60)
                return fetch_yelp_business_by_phone(phone, retry_count + 1, max_retries)
            else:
                print(f"ERROR: Max retries reached for phone {phone} due to rate limiting")
                return 'RATE_LIMITED'  # Special return value to indicate rate limiting
        else:
            print(f"Error for phone {phone}: {response.status_code}")
            return None
    except Exception as e:
        print(f"Exception for phone {phone}: {str(e)}")
        return None

def load_no_match_phones(output_dir='.'):
    """
    Load phone numbers that previously had no matches from Yelp.
    This prevents re-querying them in future runs.
    
    Returns:
        set: Set of phone numbers with no matches
    """
    no_match_file = os.path.join(output_dir, 'yelp_no_matches.txt')
    no_match_phones = set()
    
    if os.path.exists(no_match_file):
        with open(no_match_file, 'r') as f:
            no_match_phones = set(line.strip() for line in f if line.strip())
        print(f"Loaded {len(no_match_phones)} phone numbers with no previous matches")
    
    return no_match_phones

def save_no_match_phone(phone, output_dir='.'):
    """
    Save a phone number that had no match to a tracking file.
    
    Args:
        phone (str): Phone number with no match
        output_dir (str): Directory to save the file
    """
    no_match_file = os.path.join(output_dir, 'yelp_no_matches.txt')
    with open(no_match_file, 'a') as f:
        f.write(f"{phone}\n")

def load_existing_results(output_dir='.'):
    """
    Load all existing batch results and the main output file to determine 
    which phone numbers have already been processed.
    
    Returns:
        tuple: (set of processed phones, DataFrame of all existing results, set of no-match phones)
    """
    processed_phones = set()
    all_results = []
    
    # Check for main output file
    main_file = os.path.join(output_dir, 'yelp_businesses_all.csv')
    if os.path.exists(main_file):
        print(f"Loading existing results from {main_file}")
        df = pd.read_csv(main_file)
        # Convert to string to ensure consistent data types
        processed_phones.update(str(phone) for phone in df['Original_Phone'].dropna().unique())
        all_results.append(df)
    
    # Check for batch files
    batch_files = glob.glob(os.path.join(output_dir, 'yelp_results_batch_*.csv'))
    if batch_files:
        print(f"Found {len(batch_files)} batch files")
        for batch_file in batch_files:
            df = pd.read_csv(batch_file)
            # Convert to string to ensure consistent data types
            processed_phones.update(str(phone) for phone in df['Original_Phone'].dropna().unique())
            all_results.append(df)
    
    # Load no-match phones
    no_match_phones = load_no_match_phones(output_dir)
    
    # Combine all results and remove duplicates
    if all_results:
        combined_df = pd.concat(all_results, ignore_index=True)
        combined_df = combined_df.drop_duplicates(subset=['Original_Phone', 'Name'], keep='first')
        return processed_phones, combined_df, no_match_phones
    
    return processed_phones, pd.DataFrame(), no_match_phones

def process_phone_numbers(phone_numbers, batch_size=50, delay=0.5, max_calls_this_run=5000, 
                         output_dir='.', resume=True, max_consecutive_rate_limits=3):
    """
    Process a list of phone numbers and query Yelp API for each
    
    Args:
        phone_numbers (list): List of phone numbers to process
        batch_size (int): Number of phone numbers to process before saving interim results
        delay (float): Delay in seconds between API calls to avoid rate limiting
        max_calls_this_run (int): Maximum number of API calls to make this run
        output_dir (str): Directory to save results
        resume (bool): Whether to resume from existing results
        max_consecutive_rate_limits (int): Stop if this many consecutive rate limits occur
    
    Returns:
        tuple: (DataFrame of results, number of API calls made, hit_limit flag)
    """
    all_results = []
    api_calls_made = 0
    hit_limit = False
    no_match_count = 0
    consecutive_rate_limits = 0  # Track consecutive rate limit errors
    
    # Load existing results if resume is enabled
    if resume:
        processed_phones, existing_df, no_match_phones = load_existing_results(output_dir)
        if not existing_df.empty:
            all_results = existing_df.to_dict('records')
            print(f"Resuming: {len(processed_phones)} phone numbers already processed")
        else:
            processed_phones = set()
        print(f"Skipping: {len(no_match_phones)} phone numbers with no previous matches")
    else:
        processed_phones = set()
        no_match_phones = set()
        print("Starting fresh (RESUME=False) - ignoring existing results")
    
    # Ensure phone_numbers are all strings for consistent comparison
    phone_numbers = [str(p) for p in phone_numbers]
    
    # Filter out already processed phone numbers AND no-match phones
    phones_to_process = [p for p in phone_numbers if p not in processed_phones and p not in no_match_phones]
    
    # CORRECTED: Calculate actual skipped from the input list, not from loaded sets
    actually_skipped = len(phone_numbers) - len(phones_to_process)
    
    # Debug information
    print(f"\n=== DEBUG INFO ===")
    print(f"Total input phones: {len(phone_numbers)}")
    print(f"Loaded processed_phones: {len(processed_phones)}")
    print(f"Loaded no_match_phones: {len(no_match_phones)}")
    print(f"Phones after filtering: {len(phones_to_process)}")
    print(f"Actually skipped: {actually_skipped}")
    
    if len(phones_to_process) == len(phone_numbers) and len(processed_phones) > 0:
        print("\n*** WARNING: Filter may not be working correctly ***")
        print(f"Sample from phone_numbers: {phone_numbers[:3]}")
        print(f"Sample from processed_phones: {list(processed_phones)[:3]}")
        print(f"Type of phone_numbers[0]: {type(phone_numbers[0])}")
        print(f"Type of processed_phones sample: {type(list(processed_phones)[0]) if processed_phones else 'N/A'}")
    
    print(f"\nProcessing {len(phones_to_process)} new phone numbers")
    print(f"Skipping {actually_skipped} already processed")
    
    # Limit to max calls for this run
    if len(phones_to_process) > max_calls_this_run:
        print(f"\nNOTE: Limiting to {max_calls_this_run} calls this run to stay within monthly limit")
        print(f"Remaining {len(phones_to_process) - max_calls_this_run} phones will be processed in next run")
        phones_to_process = phones_to_process[:max_calls_this_run]
        hit_limit = True
    
    if len(phones_to_process) == 0:
        print("All phone numbers have been processed!")
        return pd.DataFrame(all_results), 0, False
    
    for i, phone in enumerate(tqdm(phones_to_process, desc="Processing phones")):
        # Fetch data from Yelp API
        response_data = fetch_yelp_business_by_phone(phone)
        api_calls_made += 1
        
        # Check if we hit rate limiting
        if response_data == 'RATE_LIMITED':
            consecutive_rate_limits += 1
            print(f"\nRATE LIMIT ERROR #{consecutive_rate_limits} (consecutive)")
            
            if consecutive_rate_limits >= max_consecutive_rate_limits:
                print(f"\n{'='*60}")
                print(f"STOPPING: Hit {consecutive_rate_limits} consecutive rate limits!")
                print(f"{'='*60}")
                print(f"API calls made: {api_calls_made}")
                print(f"Phones processed: {i}")
                print(f"\nYou've likely hit your DAILY API limit.")
                print(f"\nYour progress has been saved. Next steps:")
                print(f"1. Wait for your API limit to reset (usually midnight PST)")
                print(f"2. Check your API usage at: https://www.yelp.com/developers/v3/manage_app")
                print(f"3. Re-run this notebook - it will resume from where it stopped")
                print(f"{'='*60}\n")
                
                # Save all progress before stopping
                if len(all_results) > 0:
                    emergency_save = pd.DataFrame(all_results)
                    main_output = os.path.join(output_dir, 'yelp_businesses_all.csv')
                    emergency_save.to_csv(main_output, index=False)
                    print(f"Emergency save completed: {main_output}")
                
                return pd.DataFrame(all_results), api_calls_made, True
            
            # Add to no-match list and continue
            save_no_match_phone(phone, output_dir)
            no_match_count += 1
            all_results.append({
                'Original_Phone': phone,
                'Name': None,
                'Rating': None,
                'Review_Count': None,
                'Address': None,
                'City': None,
                'State': None,
                'Zip_Code': None,
                'Phone': None,
                'Categories': None,
                'Latitude': None,
                'Longitude': None,
                'Price': None,
                'Is_Closed': None,
                'URL': None
            })
        # If we got valid data and there are businesses
        elif response_data and response_data.get('businesses'):
            consecutive_rate_limits = 0  # Reset counter on success
            for business in response_data['businesses']:
                result = {
                    'Original_Phone': phone,
                    'Name': business.get('name'),
                    'Rating': business.get('rating'),
                    'Review_Count': business.get('review_count'),
                    'Address': ', '.join(business.get('location', {}).get('display_address', [])),
                    'City': business.get('location', {}).get('city'),
                    'State': business.get('location', {}).get('state'),
                    'Zip_Code': business.get('location', {}).get('zip_code'),
                    'Phone': business.get('display_phone'),
                    'Categories': ', '.join([category['title'] for category in business.get('categories', [])]),
                    'Latitude': business.get('coordinates', {}).get('latitude'),
                    'Longitude': business.get('coordinates', {}).get('longitude'),
                    'Price': business.get('price', 'N/A'),
                    'Is_Closed': business.get('is_closed'),
                    'URL': business.get('url')
                }
                all_results.append(result)
        else:
            consecutive_rate_limits = 0  # Reset counter on any non-rate-limit response
            # No match found - save to no-match tracking file
            save_no_match_phone(phone, output_dir)
            no_match_count += 1
            # Still add a record to the results for completeness (optional)
            all_results.append({
                'Original_Phone': phone,
                'Name': None,
                'Rating': None,
                'Review_Count': None,
                'Address': None,
                'City': None,
                'State': None,
                'Zip_Code': None,
                'Phone': None,
                'Categories': None,
                'Latitude': None,
                'Longitude': None,
                'Price': None,
                'Is_Closed': None,
                'URL': None
            })
        
        # Save interim results every batch_size records
        if (i+1) % batch_size == 0:
            interim_df = pd.DataFrame(all_results)
            batch_num = (len(processed_phones) + i + 1) // batch_size
            interim_filename = os.path.join(output_dir, f'yelp_results_batch_{batch_num}.csv')
            interim_df.to_csv(interim_filename, index=False)
            print(f"\nSaved {interim_filename} | Total processed: {actually_skipped + i + 1} | API calls: {api_calls_made} | No matches: {no_match_count}")
        
        # Add delay to avoid rate limiting (Yelp typically allows ~5 calls/second)
        time.sleep(delay)
    
    # Save final batch if there are unsaved results
    if len(all_results) > 0:
        final_df = pd.DataFrame(all_results)
        # Save as main output file to ensure nothing is lost
        main_output = os.path.join(output_dir, 'yelp_businesses_all.csv')
        final_df.to_csv(main_output, index=False)
        print(f"\nSaved final results to {main_output}")
        print(f"This run: {no_match_count} phone numbers had no matches (saved to yelp_no_matches.txt)")
    else:
        final_df = pd.DataFrame(all_results)
    
    return final_df, api_calls_made, hit_limit
```

```python
# ==========================================
# DATA ANALYSIS
# ==========================================
# This analyzes the complete dataset (all results from all runs)

# Analyze the categories of businesses found
if 'Categories' in yelp_df.columns and yelp_df['Categories'].notna().any():
    # Split categories and flatten the list
    all_categories = []
    for categories in yelp_df['Categories'].dropna():
        if categories:  # Check if the string isn't empty
            all_categories.extend([cat.strip() for cat in categories.split(',')])
    
    # Count occurrences of each category
    category_counts = pd.Series(all_categories).value_counts().reset_index()
    category_counts.columns = ['Category', 'Count']
    
    # Display top categories
    print("\nTop Business Categories:")
    display(category_counts.head(15))
    
    # Optionally create a bar chart
    try:
        import matplotlib.pyplot as plt
        plt.figure(figsize=(10, 6))
        top_n = 10
        plt.bar(category_counts['Category'][:top_n], category_counts['Count'][:top_n])
        plt.xticks(rotation=45, ha='right')
        plt.title(f'Top {top_n} Business Categories')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Could not create plot: {str(e)}")

# Analyze locations (states and cities)
if 'State' in yelp_df.columns and yelp_df['State'].notna().any():
    # Count businesses by state
    state_counts = yelp_df['State'].value_counts().reset_index()
    state_counts.columns = ['State', 'Count']
    
    print("\nBusinesses by State:")
    display(state_counts)
    
    # Get top cities
    if 'City' in yelp_df.columns and yelp_df['City'].notna().any():
        city_counts = yelp_df.groupby(['State', 'City']).size().reset_index(name='Count')
        city_counts = city_counts.sort_values('Count', ascending=False)
        
        print("\nTop Cities with Businesses:")
        display(city_counts.head(15))
```
