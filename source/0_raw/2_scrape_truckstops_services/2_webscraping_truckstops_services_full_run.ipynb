```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import time
import os
import json

# Load the dataframe
df = pd.read_csv(r'..\..\..\data\1_derived\scrape_truckstops_services\01_webscraping_truckstops_services_seed_urls.csv')
df
```

```python

# No state filter - process all states
# df = df[df['state'].isin(['Arizona', 'California', 'Utah', 'Nevada'])]
df = df.reset_index(drop=True)
df 
```

```python

print(f"Total rows to process: {len(df)}")

# Initialize additional fields we know about (Chain, Latitude, Longitude, LD_URL, LD_Latitude, LD_Longitude)
additional_fields = ['Chain', 'Latitude', 'Longitude', 'LD_URL', 'LD_Latitude', 'LD_Longitude']
for field in additional_fields:
    df[field] = None

# Thread-safe counter and lock for progress tracking
progress_counter = {'count': 0}
progress_lock = Lock()

# Function to extract data from a URL (now thread-safe)
def extract_locdetinfo_data(url, row_index=None, total_rows=None):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')
        data = {}

        # Find the leftcol div and extract all locdetinfo data
        leftcol = soup.find('div', id='leftcol')
        if leftcol:
            locdetinfo_divs = leftcol.find_all('div', class_='locdetinfo')
            for div in locdetinfo_divs:
                div_text = div.get_text().strip()
                match = re.match(r'([^:]+):\s*(.+)', div_text)
                if match:
                    field_name = match.group(1).strip()
                    field_value = match.group(2).strip()
                    data[field_name] = field_value

        # Extract Chain information from popb_rotate div
        popb_rotate_div = soup.find('div', class_='popb_rotate')
        if popb_rotate_div:
            img_tag = popb_rotate_div.find('img')
            if img_tag and img_tag.get('src'):
                img_src = img_tag.get('src')
                filename = img_src.split('/')[-1]
                chain_name = filename.replace('.jpg', '').replace('.png', '').replace('logo', '').strip()
                data['Chain'] = chain_name

        # Extract coordinates from Google Maps link
        maps_link = soup.find('a', href=lambda x: x and 'google.com/maps/place' in x if x else False)
        if maps_link:
            href = maps_link.get('href')
            coord_match = re.search(r'place/(-?\d+\.?\d*),(-?\d+\.?\d*)', href)
            if coord_match:
                data['Latitude'] = coord_match.group(1)
                data['Longitude'] = coord_match.group(2)

        # Extract info from <script type="application/ld+json">
        ld_json_script = soup.find('script', type='application/ld+json')
        if ld_json_script:
            try:
                ld_json = json.loads(ld_json_script.string)
                # Extract url, geo.latitude, geo.longitude if present
                if 'url' in ld_json:
                    data['LD_URL'] = ld_json['url']
                if 'geo' in ld_json and isinstance(ld_json['geo'], dict):
                    if 'latitude' in ld_json['geo']:
                        data['LD_Latitude'] = ld_json['geo']['latitude']
                    if 'longitude' in ld_json['geo']:
                        data['LD_Longitude'] = ld_json['geo']['longitude']
            except Exception as e:
                print(f"Error parsing ld+json for {url}: {e}")

        return data

    except Exception as e:
        # Suppress individual errors to keep output clean
        return {}

# Function to process a single row (used by concurrent processing)
def process_row(row_data):
    index, row = row_data
    url = row['full_url']
    name = row['name']
    
    # Extract data
    extracted_data = extract_locdetinfo_data(url)
    
    # Add small random delay to be respectful (0.2-0.5 seconds per request)
    time.sleep(0.3)
    
    # Update progress counter
    with progress_lock:
        progress_counter['count'] += 1
        current = progress_counter['count']
    
    return index, name, extracted_data, current

# CONCURRENT BATCH PROCESSING - Much faster!
batch_size = 1000
total_rows = len(df)
processed_dfs = []
all_new_columns = set(additional_fields)
max_workers = 10  # Process 10 URLs concurrently (adjust based on your needs)

print(f"Processing {total_rows} rows in batches of {batch_size}")
print(f"Using {max_workers} concurrent workers for MUCH faster processing")
print("=" * 80)

for batch_num in range(0, total_rows, batch_size):
    batch_end = min(batch_num + batch_size, total_rows)
    current_batch_num = (batch_num // batch_size) + 1
    total_batches = (total_rows - 1) // batch_size + 1

    print(f"\nBATCH {current_batch_num}/{total_batches}: Processing rows {batch_num} to {batch_end-1}")
    print("-" * 60)
    
    batch_start_time = time.time()
    batch_df = df.iloc[batch_num:batch_end].copy()
    
    # Prepare data for concurrent processing
    row_data_list = list(batch_df.iterrows())
    
    # Process rows concurrently
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        # Submit all tasks
        future_to_row = {executor.submit(process_row, row_data): row_data for row_data in row_data_list}
        
        # Process completed tasks
        completed = 0
        for future in as_completed(future_to_row):
            try:
                index, name, extracted_data, current = future.result()
                completed += 1
                
                # Track new columns
                for field_name in extracted_data.keys():
                    if field_name not in all_new_columns:
                        all_new_columns.add(field_name)
                        print(f"    New field discovered: {field_name}")
                
                # Update dataframe
                for field_name, field_value in extracted_data.items():
                    if field_name not in batch_df.columns:
                        batch_df[field_name] = None
                    batch_df.loc[index, field_name] = field_value
                
                # Print progress every 50 rows
                if completed % 50 == 0:
                    elapsed = time.time() - batch_start_time
                    rate = completed / elapsed if elapsed > 0 else 0
                    eta = (len(batch_df) - completed) / rate if rate > 0 else 0
                    print(f"    Progress: {completed}/{len(batch_df)} | Rate: {rate:.1f} rows/sec | ETA: {eta/60:.1f} min")
                    
            except Exception as e:
                print(f"    Error processing row: {e}")

    for col in all_new_columns:
        if col not in batch_df.columns:
            batch_df[col] = None

    batch_filename = f'02_all_states_truckstops_services_batch_{current_batch_num:02d}_of_16.csv'
    batch_path = os.path.join(r'..\..\..\temp\batch_scrape_truckstops_services', batch_filename)
    batch_df.to_csv(batch_path, index=False)
    
    batch_time = time.time() - batch_start_time
    print(f"  Batch {current_batch_num} saved as: {batch_filename}")
    print(f"  BATCH {current_batch_num} COMPLETED: {len(batch_df)} rows in {batch_time/60:.1f} minutes ({len(batch_df)/batch_time:.1f} rows/sec)")

    processed_dfs.append(batch_df)

print("\n" + "=" * 80)
print("COMBINING ALL BATCHES...")

for i, batch_df in enumerate(processed_dfs):
    for col in all_new_columns:
        if col not in batch_df.columns:
            batch_df[col] = None

final_df = pd.concat(processed_dfs, ignore_index=True)

# Save as CSV - labeled as ALL STATES version
output_path_csv = r'..\..\..\output\0_raw\1_truckstops_services_webscraped.csv'
final_df.to_csv(output_path_csv, index=False)

# Save as Parquet (more efficient for large datasets) - labeled as ALL STATES version
# output_path_parquet = r'..\..\..\output\0_raw\1_truckstops_services_webscraped.parquet'
# final_df.to_parquet(output_path_parquet, index=False, compression='snappy')

print(f"SUCCESS! All {total_rows} rows processed and saved to:")
print(f"  - CSV: 1_truckstops_services_webscraped.csv ({os.path.getsize(output_path_csv) / 1024 / 1024:.2f} MB)")
print(f"  - Parquet: 1_truckstops_services_webscraped.parquet ({os.path.getsize(output_path_parquet) / 1024 / 1024:.2f} MB)")
print(f"  - Note: This file contains ALL states (not filtered)")
print(f"Final dataset shape: {final_df.shape}")
print(f"Total columns: {len(final_df.columns)}")
print(f"New columns discovered: {sorted(list(all_new_columns - set(additional_fields)))}")

print("\n" + "=" * 80)
print("FINAL RESULTS SUMMARY:")
print("-" * 40)
for col in sorted(final_df.columns):
    non_null_count = final_df[col].notna().sum()
    percentage = (non_null_count / len(final_df)) * 100
    print(f"{col}: {non_null_count}/{len(final_df)} ({percentage:.1f}%)")

print(f"\nFirst 5 rows of final dataset:")
final_df.head()
```
