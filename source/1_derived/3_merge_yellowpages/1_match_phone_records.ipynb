# 1 Match Phone Records
Cross-dataset phone matching: match OCR truck stop phone numbers against Yellow Pages scraped records.

Inputs:
- data/1_derived/3_scrape_yelp_truckstops/5_place_changes.csv  (OCR dataset with prior Yelp match columns — final Yelp pipeline output)
- output/0_raw/3_yellowpages_webscraped.csv  (Yellow Pages scraped data; ORIGINAL_PHONE column)

Output: data/1_derived/4_scrape_yellowpages/1_yellowpages_phone_matched.csv

```python
from pathlib import Path
import re
import pandas as pd


def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'data').exists() and (candidate / 'source').exists() and (candidate / 'temp').exists():
            return candidate
    return start


PROJECT_ROOT = find_project_root(Path.cwd())
YP_DIR = PROJECT_ROOT / 'data' / '1_derived' / '4_scrape_yellowpages'
YP_DIR.mkdir(parents=True, exist_ok=True)
# Read from TruckStops merge output (OCR data + scraped match columns)
OCR_FILE = PROJECT_ROOT / 'temp' / '1_derived' / '1_merge_station_records.csv'
YP_SCRAPED_FILE = PROJECT_ROOT / 'output' / '0_raw' / '3_yellowpages_webscraped.csv'
OUT_FILE = YP_DIR / '1_yellowpages_phone_matched.csv'


def normalize_phone(value):
    if pd.isna(value):
        return None
    digits = re.sub(r'\D', '', str(value))
    if len(digits) < 10:
        return None
    if len(digits) == 10:
        return '1' + digits
    if len(digits) == 11 and digits.startswith('1'):
        return digits
    return '1' + digits[-10:]


# Load OCR dataset (main truck stop records with prior Yelp match columns)
df_ocr = pd.read_csv(OCR_FILE, low_memory=False)
print(f'OCR records loaded: {len(df_ocr):,} rows')

# Normalize OCR phones early (needed for both Yelp and YP matching)
df_ocr['_norm_phone'] = df_ocr['phone'].apply(normalize_phone)

# Load Yelp scraped data and create Yelp phone match columns on OCR data
YELP_FILE = PROJECT_ROOT / 'output' / '0_raw' / '2_yelp_webscraped.csv'
df_yelp = pd.read_csv(YELP_FILE, low_memory=False)
_yelp_phone_col = 'Original_Phone' if 'Original_Phone' in df_yelp.columns else 'Phone'
df_yelp['_norm_phone'] = df_yelp[_yelp_phone_col].apply(normalize_phone)
_yelp_valid = df_yelp['Name'].notna() if 'Name' in df_yelp.columns else pd.Series([True]*len(df_yelp))
_yelp_phone_to_ids = {}
for _idx, _phone in df_yelp.loc[_yelp_valid, '_norm_phone'].items():
    if _phone:
        _yelp_phone_to_ids.setdefault(_phone, []).append(int(_idx))
df_ocr['Phone_Yelp_matches_row_ids'] = df_ocr['_norm_phone'].apply(
    lambda p: _yelp_phone_to_ids.get(p, []) if p else [])
df_ocr['Yelp_phone_match_rate'] = df_ocr['Phone_Yelp_matches_row_ids'].apply(lambda x: len(x) > 0 if isinstance(x, list) else False)
print(f'Yelp phone matches: {df_ocr["Yelp_phone_match_rate"].sum():,} / {len(df_ocr):,}')

# Load Yellow Pages scraped data
df_yp = pd.read_csv(YP_SCRAPED_FILE, low_memory=False)
print(f'YP scraped records loaded: {len(df_yp):,} rows')

# Normalize phone columns in both datasets
df_yp['_norm_phone'] = df_yp['ORIGINAL_PHONE'].apply(normalize_phone)

# Build lookup: normalized phone → list of row IDs in YP dataset
# Only count rows where a business was actually found (BUSINESS_NAME not null)
yp_found_mask = df_yp['BUSINESS_NAME'].notna()
phone_to_yp_row_ids = {}
phone_to_yp_found_ids = {}
for idx, row in df_yp.iterrows():
    p = row['_norm_phone']
    if p:
        phone_to_yp_row_ids.setdefault(p, []).append(int(idx))
        if pd.notna(row['BUSINESS_NAME']):
            phone_to_yp_found_ids.setdefault(p, []).append(int(idx))

print(f'Unique phones indexed in YP dataset: {len(phone_to_yp_row_ids):,}')
print(f'Unique phones with a found business: {len(phone_to_yp_found_ids):,}')

# Match each OCR phone against the YP lookup
df_ocr['Phone_Yellowpages_matches_row_ids'] = df_ocr['_norm_phone'].apply(
    lambda p: phone_to_yp_row_ids.get(p, []) if p else []
)
# Yellowbook_phone_match_rate: True only when YP returned an actual business result
df_ocr['Yellowbook_phone_match_rate'] = df_ocr['_norm_phone'].apply(
    lambda p: bool(phone_to_yp_found_ids.get(p)) if p else False
)

df_ocr = df_ocr.drop(columns=['_norm_phone'])

matched = (df_ocr['Phone_Yellowpages_matches_row_ids'].apply(len) > 0).sum()
found = df_ocr['Yellowbook_phone_match_rate'].sum()
print(f'\nOCR rows with any YP phone match : {matched:,} ({matched/len(df_ocr)*100:.1f}%)')
print(f'OCR rows with a found YP business: {found:,} ({found/len(df_ocr)*100:.1f}%)')


# Create match-rate columns from TruckStops matching (needed by downstream notebooks)
import ast as _ast

def _parse_ids(value):
    if isinstance(value, list):
        return set(value)
    if pd.isna(value):
        return set()
    if isinstance(value, str):
        try:
            parsed = _ast.literal_eval(value)
            if isinstance(parsed, list):
                return set(parsed)
        except Exception:
            pass
    return set()

def _has_match(value):
    return len(_parse_ids(value)) > 0

# Scraped_phone_match_rate: True if phone matched in scraped data
if 'phone_scraped_matches_row_ids' in df_ocr.columns and 'Scraped_phone_match_rate' not in df_ocr.columns:
    df_ocr['Scraped_phone_match_rate'] = df_ocr['phone_scraped_matches_row_ids'].apply(_has_match)
    print(f'Created Scraped_phone_match_rate: {df_ocr["Scraped_phone_match_rate"].sum():,} matches')

# Scraped_zipcode_to_label_match_rate: multi-criteria matching score
_id_cols = ['ZIP_scraped_matches_row_ids', 'State_scraped_matches_row_ids',
            'City_scraped_matches_row_ids', 'Exit_scraped_matches_row_ids',
            'Road_scraped_matches_row_ids', 'Label_scraped_matches_row_ids',
            'Chain_scraped_matches_row_ids']
if all(c in df_ocr.columns for c in _id_cols) and 'Scraped_zipcode_to_label_match_rate' not in df_ocr.columns:
    for c in _id_cols:
        df_ocr[c + '_parsed'] = df_ocr[c].apply(_parse_ids)

    def _determine_match_success(row):
        ms = row['ZIP_scraped_matches_row_ids_parsed'] | row['State_scraped_matches_row_ids_parsed']
        if not ms:
            return "0/6 successful match"
        ce = row['City_scraped_matches_row_ids_parsed'] | row['Exit_scraped_matches_row_ids_parsed']
        ms = ms & ce if ce else ms
        if not ms:
            return "2/6 successful match"
        rd = row['Road_scraped_matches_row_ids_parsed']
        ms = ms & rd if rd else ms
        if not ms:
            return "4/6 successful match"
        lb = row['Label_scraped_matches_row_ids_parsed']
        ms = ms & lb if lb else ms
        if not ms:
            return "5/6 successful match"
        ch = row['Chain_scraped_matches_row_ids_parsed']
        ms = ms & ch if ch else ms
        if not ms:
            return "6/6 successful match"
        return "7/6 successful match"

    df_ocr['Scraped_zipcode_to_label_match_rate'] = df_ocr.apply(_determine_match_success, axis=1)
    df_ocr = df_ocr.drop(columns=[c + '_parsed' for c in _id_cols])
    print(f'Created Scraped_zipcode_to_label_match_rate')
    print(df_ocr['Scraped_zipcode_to_label_match_rate'].value_counts().sort_index())


# Create place_identifier(year) by following the target year sequence
# Algorithm: scan rows in order (data is sorted by phone, year). When a row's year
# matches a position in the target sequence, start a group. Continue the group while
# subsequent rows match the next expected year. Break on mismatch.
_target_sequence = [2006, 2007, 2008, 2014, 2015, 2016]
if 'year' in df_ocr.columns and 'place_identifier(year)' not in df_ocr.columns:
    _n = 1
    _group_ids = [None] * len(df_ocr)
    _i = 0
    while _i < len(df_ocr):
        _current_year = df_ocr.at[_i, 'year']
        if _current_year in _target_sequence:
            _seq_idx = _target_sequence.index(_current_year)
            _group_ids[_i] = _n
            _seq_idx += 1
            _i += 1
            while _i < len(df_ocr) and _seq_idx < len(_target_sequence):
                _current_year = df_ocr.at[_i, 'year']
                if _current_year == _target_sequence[_seq_idx]:
                    _group_ids[_i] = _n
                    _seq_idx += 1
                    _i += 1
                else:
                    break
            _n += 1
        else:
            _i += 1
    df_ocr['place_identifier(year)'] = _group_ids
    _assigned = df_ocr['place_identifier(year)'].notna().sum()
    print(f'Created place_identifier(year): {_assigned:,} rows assigned ({_n-1} places)')

df_ocr.to_csv(OUT_FILE, index=False)
print(f'\nSaved: {OUT_FILE}')
df_ocr.head(3)
```
