# 7 Apply Manual Fixes
Apply manually-reviewed coordinates from reference_code/8.ipynb using corrected place identifiers.

The original manual fixes used place_identifier(year) values from an older dataset (10_supplementary.csv).
This notebook maps those old IDs to our current IDs via phone+year matching, then applies the
Manual_Lat, Manual_Long, and Match_Comments to Final_Lat/Final_Long.

Input:  data/1_derived/5_geocode_truck_stops/7_final_truck_stops.csv
        data/0_raw/reference/10_supplementary.csv
Output: data/1_derived/5_geocode_truck_stops/8_final_truck_stops_manual.csv

```python
from pathlib import Path
import pandas as pd


def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'data').exists() and (candidate / 'source').exists() and (candidate / 'temp').exists():
            return candidate
    return start


PROJECT_ROOT = find_project_root(Path.cwd())
FINAL_FILE = PROJECT_ROOT / 'data' / '1_derived' / '5_geocode_truck_stops' / '7_final_truck_stops.csv'
SUPP_FILE  = PROJECT_ROOT / 'data' / '0_raw' / 'reference' / '10_supplementary.csv'
REVIEW_FILE = PROJECT_ROOT / 'data' / '1_derived' / '5_geocode_truck_stops' / '3_manual_review_ready.csv'

df = pd.read_csv(FINAL_FILE, low_memory=False)
supp = pd.read_csv(SUPP_FILE, low_memory=False)
review = pd.read_csv(REVIEW_FILE, low_memory=False)
print(f'Final:  {len(df):,} rows')
print(f'Supp:   {len(supp):,} rows')
print(f'Review: {len(review):,} rows')
```

```python
# ── Build old→new place_identifier(year) mapping via phone+year ──
old_ids = list(manual_fixes.keys())

# Get identifying info from supplementary (old IDs)
supp_subset = supp[supp['place_identifier(year)'].isin(old_ids)][
    ['place_identifier(year)', 'phone', 'year']
].drop_duplicates('place_identifier(year)')

# Match to our data using phone+year
id_map = {}  # old_id -> new_id
for _, row in supp_subset.iterrows():
    old_id = row['place_identifier(year)']
    match = review[(review['phone'] == row['phone']) & (review['year'] == row['year'])]
    if len(match) >= 1:
        new_id = match['place_identifier(year)'].iloc[0]
        if pd.notna(new_id) and pd.notna(old_id):
            id_map[int(old_id)] = int(new_id)
        else:
            print(f'WARNING: NaN place_identifier for old_id={old_id} -> new_id={new_id}')
    else:
        print(f'WARNING: No match for old_id={old_id} (phone={row["phone"]}, year={row["year"]})')

print(f'Mapped {len(id_map)} of {len(old_ids)} old IDs to new IDs')
print()
for old_id, new_id in sorted(id_map.items()):
    print(f'  {old_id:>4} -> {new_id}')
```

```python
# ── Apply manual fixes to Final_Lat / Final_Long / Match_Comments ──
# Ensure Match_Comments is string type (may be all-NaN float64)
df['Match_Comments'] = df['Match_Comments'].fillna('').astype(str)

total_rows_updated = 0

for old_id, fixes in manual_fixes.items():
    if old_id not in id_map:
        print(f'SKIP old_id={old_id} (no mapping)')
        continue
    new_id = id_map[old_id]
    mask = df['place_identifier(year)'] == new_id
    n_rows = mask.sum()

    # Apply Manual_Lat/Long -> Final_Lat/Long
    if 'Manual_Lat' in fixes and 'Manual_Long' in fixes:
        df.loc[mask, 'Final_Lat'] = fixes['Manual_Lat']
        df.loc[mask, 'Final_Long'] = fixes['Manual_Long']

    # Apply Match_Comments
    if 'Match_Comments' in fixes:
        df.loc[mask, 'Match_Comments'] = fixes['Match_Comments']

    total_rows_updated += n_rows
    label = df.loc[mask, 'label'].iloc[0] if n_rows > 0 else '?'
    coord = f"({fixes.get('Manual_Lat', '-')}, {fixes.get('Manual_Long', '-')})" if 'Manual_Lat' in fixes else '(comment only)'
    print(f'  old {old_id:>4} -> new {new_id:>5} | {n_rows} rows | {label[:40]:<40} | {coord}')

print(f'\nTotal rows updated: {total_rows_updated}')
```

```python
# ── Completion rate after manual fixes ──
total = len(df)
has_both = df['Final_Lat'].notna() & df['Final_Long'].notna()
print(f'Total rows          : {total:,}')
print(f'Has coordinates     : {has_both.sum():,}  ({has_both.sum()/total*100:.2f}%)')
print(f'Missing coordinates : {(~has_both).sum():,}  ({(~has_both).sum()/total*100:.2f}%)')

# Save to separate output file
OUT_FILE = PROJECT_ROOT / 'data' / '1_derived' / '5_geocode_truck_stops' / '8_final_truck_stops_manual.csv'
df.to_csv(OUT_FILE, index=False)
print(f'\nSaved: {OUT_FILE}')
```
