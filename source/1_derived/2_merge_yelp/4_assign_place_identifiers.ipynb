# 4 Assign Place Identifiers
Assign place_identifier(year) using reference year-sequence logic.

Input: data/1_derived/3_scrape_yelp_truckstops/3_reordered_rows.csv
Output: data/1_derived/3_scrape_yelp_truckstops/4_place_identifier.csv

```python
from pathlib import Path
import pandas as pd

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'data').exists() and (candidate / 'source').exists() and (candidate / 'temp').exists():
            return candidate
    return start

PROJECT_ROOT = find_project_root(Path.cwd())
OUT_DIR = PROJECT_ROOT / 'data' / '1_derived' / '3_scrape_yelp_truckstops'
IN_FILE = OUT_DIR / '3_reordered_rows.csv'
OUT_FILE = OUT_DIR / '4_place_identifier.csv'

df = pd.read_csv(IN_FILE)
target_years = [2006, 2007, 2008, 2014, 2015, 2016]

if 'year' in df.columns:
    df = df.sort_values(by=['year']).reset_index(drop=True)
    df['place_identifier(year)'] = None
    place_counter = 1
    i = 0
    while i < len(df):
        if i + len(target_years) <= len(df):
            year_sequence = df.iloc[i:i + len(target_years)]['year'].tolist()
            if year_sequence == target_years:
                df.loc[i:i + len(target_years) - 1, 'place_identifier(year)'] = place_counter
                place_counter += 1
                i += len(target_years)
                continue
        i += 1
    print(f"Assigned place_identifier(year) for {df['place_identifier(year)'].notna().sum()} rows.")
else:
    df['place_identifier(year)'] = None
    print('No year column found; created empty place_identifier(year).')

df.to_csv(OUT_FILE, index=False)
print(f'Saved: {OUT_FILE}')
```
