# 1 Match Phone Records
Build phone-normalized matching indices from Yelp records.

Input: data/1_derived/3_scrape_yelp_truckstops/yelp_businesses_all.csv
Output: data/1_derived/3_scrape_yelp_truckstops/1_phone_matches.csv

```python
from pathlib import Path
import re
import pandas as pd

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'data').exists() and (candidate / 'source').exists() and (candidate / 'temp').exists():
            return candidate
    return start

PROJECT_ROOT = find_project_root(Path.cwd())
OUT_DIR = PROJECT_ROOT / 'data' / '1_derived' / '3_scrape_yelp_truckstops'
IN_FILE = PROJECT_ROOT / 'output' / '0_raw' / '2_yelp_webscraped.csv'
OUT_FILE = OUT_DIR / '1_phone_matches.csv'

def normalize_phone(value):
    if pd.isna(value):
        return None
    digits = re.sub(r'\D', '', str(value))
    if len(digits) < 10:
        return None
    if len(digits) == 10:
        return '1' + digits
    if len(digits) == 11 and digits.startswith('1'):
        return digits
    return '1' + digits[-10:]

df = pd.read_csv(IN_FILE)
phone_col = 'phone' if 'phone' in df.columns else 'Original_Phone'
df['normalized_phone'] = df[phone_col].apply(normalize_phone)

valid_name_mask = df['Name'].notna() if 'Name' in df.columns else pd.Series([True] * len(df), index=df.index)
phone_to_row_ids = {}
for idx, phone in df.loc[valid_name_mask, 'normalized_phone'].items():
    if phone:
        phone_to_row_ids.setdefault(phone, []).append(int(idx))

df['Phone_Yelp_matches_row_ids'] = df['normalized_phone'].apply(lambda p: phone_to_row_ids.get(p, []) if p else [])
df.to_csv(OUT_FILE, index=False)
print(f'Saved: {OUT_FILE}')
df.head(5)
```
