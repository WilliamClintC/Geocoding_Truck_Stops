# 2 Build Match Rate Flags
Standardize match-rate columns and create Yelp_phone_match_rate.

Input: data/1_derived/3_scrape_yelp_truckstops/1_phone_matches.csv
Output: data/1_derived/3_scrape_yelp_truckstops/2_match_rates.csv

```python
from pathlib import Path
import pandas as pd
import ast

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'data').exists() and (candidate / 'source').exists() and (candidate / 'temp').exists():
            return candidate
    return start

PROJECT_ROOT = find_project_root(Path.cwd())
OUT_DIR = PROJECT_ROOT / 'data' / '1_derived' / '3_scrape_yelp_truckstops'
IN_FILE = OUT_DIR / '1_phone_matches.csv'
OUT_FILE = OUT_DIR / '2_match_rates.csv'

df = pd.read_csv(IN_FILE)

rename_map = {}
if 'Success_Match_Rate' in df.columns:
    rename_map['Success_Match_Rate'] = 'Scraped_zipcode_to_label_match_rate'
if 'Phone_Success_Match_Rate' in df.columns:
    rename_map['Phone_Success_Match_Rate'] = 'Scraped_phone_match_rate'
if rename_map:
    df = df.rename(columns=rename_map)

def has_match(value):
    if isinstance(value, list):
        return len(value) > 0
    if pd.isna(value):
        return False
    if isinstance(value, str):
        try:
            parsed = ast.literal_eval(value)
            return isinstance(parsed, list) and len(parsed) > 0
        except Exception:
            return False
    return False

df['Yelp_phone_match_rate'] = df['Phone_Yelp_matches_row_ids'].apply(has_match)
df.to_csv(OUT_FILE, index=False)
print(f'Saved: {OUT_FILE}')
print(df['Yelp_phone_match_rate'].value_counts(dropna=False))
```
