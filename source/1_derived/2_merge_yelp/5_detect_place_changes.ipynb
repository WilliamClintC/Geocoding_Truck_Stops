# 5 Detect Place Changes
Detect place changes across years using label similarity and chain matching.

Input: data/1_derived/3_scrape_yelp_truckstops/4_place_identifier.csv
Output: data/1_derived/3_scrape_yelp_truckstops/5_place_changes.csv

```python
from pathlib import Path
import re
import pandas as pd
from difflib import SequenceMatcher

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'data').exists() and (candidate / 'source').exists() and (candidate / 'temp').exists():
            return candidate
    return start

PROJECT_ROOT = find_project_root(Path.cwd())
OUT_DIR = PROJECT_ROOT / 'data' / '1_derived' / '3_scrape_yelp_truckstops'
IN_FILE = OUT_DIR / '4_place_identifier.csv'
OUT_FILE = OUT_DIR / '5_place_changes.csv'

df = pd.read_csv(IN_FILE)

common_keywords = {
    'truck', 'stop', 'travel', 'center', 'fuel', 'gas', 'station', 'service',
    'mart', 'store', 'shop', 'plaza', 'loves', 'pilot', 'ta', 'petro',
    'chevron', 'shell', 'exxon', 'mobil'
}

def extract_keywords(label):
    if pd.isna(label):
        return set(), set()
    words = re.findall(r'[a-zA-Z0-9]+', str(label).lower())
    distinctive = set()
    common = set()
    for w in words:
        if w in common_keywords:
            common.add(w)
        elif len(w) > 1:
            distinctive.add(w)
    return distinctive, common

def calculate_similarity(label1, label2):
    if pd.isna(label1) or pd.isna(label2):
        return 0.0
    d1, c1 = extract_keywords(label1)
    d2, c2 = extract_keywords(label2)
    dist_sim = len(d1.intersection(d2)) / len(d1.union(d2)) if d1.union(d2) else 0.0
    common_sim = len(c1.intersection(c2)) / len(c1.union(c2)) if c1.union(c2) else 0.0
    str_sim = SequenceMatcher(None, str(label1).lower(), str(label2).lower()).ratio()
    bonus = min(0.4, 0.2 * len(d1.intersection(d2)))
    if len(d1.intersection(d2)) >= 2:
        bonus += 0.1
    weighted = (0.4 * dist_sim) + (0.3 * str_sim) + (0.1 * common_sim) + bonus
    return min(1.0, weighted)

def check_chain_match(c1, c2):
    if pd.isna(c1) or pd.isna(c2):
        return True
    return str(c1).strip().lower() == str(c2).strip().lower()

required = {'place_identifier(year)', 'year', 'label', 'chain'}
if required.issubset(set(df.columns)):
    df['Flag_Place_Change'] = False
    df['Flag_Place_Change_Year'] = None
    df['Similarity_Score'] = None
    df['Chain_Match'] = None
    df['Previous_Year_Label'] = None
    df['Previous_Year_Chain'] = None

    valid_groups = df.dropna(subset=['place_identifier(year)']).groupby('place_identifier(year)')
    for pid, group in valid_groups:
        grp = group.sort_values('year')
        prev = None
        for idx, row in grp.iterrows():
            if prev is None:
                prev = row
                continue
            sim = calculate_similarity(prev['label'], row['label'])
            chain_match = check_chain_match(prev['chain'], row['chain'])
            changed = (sim < 0.5 and not chain_match) or (sim < 0.2)
            df.at[idx, 'Flag_Place_Change'] = bool(changed)
            df.at[idx, 'Flag_Place_Change_Year'] = int(row['year'])
            df.at[idx, 'Similarity_Score'] = float(sim)
            df.at[idx, 'Chain_Match'] = bool(chain_match)
            df.at[idx, 'Previous_Year_Label'] = prev['label']
            df.at[idx, 'Previous_Year_Chain'] = prev['chain']
            prev = row
    print('Applied place-change detection.')
else:
    for col, default in [
        ('Flag_Place_Change', False),
        ('Flag_Place_Change_Year', None),
        ('Similarity_Score', None),
        ('Chain_Match', None),
        ('Previous_Year_Label', None),
        ('Previous_Year_Chain', None)
    ]:
        df[col] = default
    print('Missing required columns for full change detection. Added empty output columns.')

df.to_csv(OUT_FILE, index=False)
print(f'Saved: {OUT_FILE}')
df.head(5)
```
