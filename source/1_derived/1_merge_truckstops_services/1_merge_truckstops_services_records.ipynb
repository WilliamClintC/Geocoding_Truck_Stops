# **Geocoding Task Notebook Explanation**

This notebook implements various matching algorithms between two datasets:
- `df1 - Main dataset with gas stations
- `df2- Scraped dataset with additional information

We'll match these datasets using several criteria:
1. Phone number matching
2. ZIP code matching
3. City name matching
4. Exit number matching
5. State matching
6. Road name matching
7. Chain name matching
8. Label text matching

Each matching will create a new column in df1 containing the row IDs of matching rows in df2.

```python
# Load the required datasets
import pandas as pd
import re
import numpy as np
import time

# Start timing
start_time = time.time()
print(f"Starting processing at: {time.strftime('%Y-%m-%d %H:%M:%S')}")
print("="*60)

# Display the first few rows of each dataset
print("Main dataset (df1) shape:", df1.shape)
print("Scraped dataset (df2) shape:", df2.shape)

# Display a sample of each dataset
display(df1.head(3))
display(df2.head(3))

# Reset the indices to ensure they're sequential
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
```

```python
# CENTRALIZED COMMON WORDS FILTER
# This comprehensive filter will be used across all matching functions for consistency

def get_comprehensive_common_words_filter():
    """
    Returns a comprehensive set of common words that should be filtered out
    across all matching algorithms to prevent false matches.
    """
    common_words = {
        # Basic English stop words
        'the', 'of', 'and', 'to', 'a', 'in', 'for', 'is', 'on', 'that', 'by', 'this', 'with',
        'i', 'you', 'it', 'at', 'be', 'or', 'an', 'are', 'as', 'be', 'been', 'by',
        'for', 'from', 'has', 'he', 'in', 'is', 'it', 'its', 'of', 'on', 'that', 'the',
        'to', 'was', 'will', 'with', 'but', 'not', 'all', 'can', 'had', 'her', 'his',
        'one', 'our', 'out', 'day', 'get', 'use', 'man', 'new', 'now', 'way', 'may', 'say',
        
        # Symbols and special characters (as words)
        '#', '&', '@', 'and', 'or', 'vs', 'inc', 'llc', 'ltd', 'corp', 'co',
        
        # Business/location related common words
        'gas', 'station', 'service', 'store', 'shop', 'center', 'mart', 'market',
        'fuel', 'food', 'auto', 'car', 'truck', 'travel', 'stop', 'plaza',
        
        # Street types and directionals
        'st', 'street', 'ave', 'avenue', 'blvd', 'boulevard', 'rd', 'road', 'ln', 'lane',
        'dr', 'drive', 'way', 'pkwy', 'parkway', 'hwy', 'highway', 'expwy', 'expressway',
        'ct', 'court', 'cir', 'circle', 'pl', 'place', 'ter', 'terrace', 'trl', 'trail',
        'route', 'rte', 'miles', 'mile', 'mi', 'km', 'exit', 'interchange',
        
        # Directionals
        'n', 's', 'e', 'w', 'north', 'south', 'east', 'west', 'ne', 'nw', 'se', 'sw',
        'northeast', 'northwest', 'southeast', 'southwest', 'northern', 'southern',
        'eastern', 'western',
        
        # Geographic terms
        'city', 'town', 'village', 'county', 'state', 'country', 'usa', 'us', 'america',
        'united', 'states',
        
        # Numbers as words (common ones that might appear in addresses)
        'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten',
        'first', 'second', 'third', 'fourth', 'fifth', 'sixth', 'seventh', 'eighth',
        'ninth', 'tenth', 'hundred', 'thousand',
        
        # Time-related words
        'am', 'pm', 'hour', 'hours', 'day', 'days', 'week', 'weeks', 'month', 'months',
        'year', 'years', 'open', 'close', 'closed',
        
        # Size/quantity descriptors
        'big', 'small', 'large', 'little', 'mini', 'super', 'mega', 'giant', 'express',
        'quick', 'fast', 'speed', 'rapid',
        
        # Common business suffixes/prefixes
        'old', 'new', 'fresh', 'best', 'good', 'great', 'quality', 'family', 'local',
        'discount', 'cheap', 'budget', 'premium', 'deluxe'
    }
    
    return common_words

def clean_ordinal_suffixes(text_str):
    """
    Removes ordinal suffixes from numbers in text.
    Examples: "66th" â†’ "66", "2nd" â†’ "2", "1st" â†’ "1"
    """
    if pd.isna(text_str):
        return ''
    
    # Remove ordinal suffixes from numbers
    cleaned = re.sub(r'(\d+)(?:st|nd|rd|th)\b', r'\1', str(text_str).lower())
    return cleaned
```

```python
# STEP 1: Match by phone number (OPTIMIZED)
print(f"\n{'='*60}")
print("STEP 1: Phone Number Matching (OPTIMIZED)")
print(f"{'='*60}")
step_start = time.time()

# Create the new column to store matching row IDs
df1['phone_scraped_matches_row_ids'] = None

# Function to clean phone numbers by removing non-numeric characters
def clean_phone_number(phone_str):
    if pd.isna(phone_str):
        return ''
    # Extract only digits from the phone number
    digits_only = re.sub(r'\D', '', str(phone_str))
    # If we have at least 7 digits, consider it a valid phone number
    if len(digits_only) >= 7:
        return digits_only
    return ''

# Clean phone numbers in df1
df1_clean_phones = df1['phone'].apply(clean_phone_number)

# Initialize the phone columns to check in df2
phone_columns = ['Phone', 'Phone 2', 'Phone 3', 'Phone 4', 'Phone 5', 'Fax']

# Create cleaned versions of all phone columns in df2
df2_clean_phones = {}
for col in phone_columns:
    if col in df2.columns:
        df2_clean_phones[col] = df2[col].apply(clean_phone_number)
    else:
        print(f"Warning: Column '{col}' not found in df2")

# OPTIMIZATION: Build reverse index - phone number to df2 row IDs
print("Building phone number index...")
phone_to_df2_rows = {}
for col, clean_phones in df2_clean_phones.items():
    for j, phone in enumerate(clean_phones):
        if phone:  # Only index non-empty phones
            if phone not in phone_to_df2_rows:
                phone_to_df2_rows[phone] = []
            phone_to_df2_rows[phone].append(j)

# Remove duplicates from the index
for phone in phone_to_df2_rows:
    phone_to_df2_rows[phone] = list(set(phone_to_df2_rows[phone]))

# Vectorized matching using the index
matches_list = []
for phone in df1_clean_phones:
    if phone and phone in phone_to_df2_rows:
        matches_list.append(str(phone_to_df2_rows[phone]))
    else:
        matches_list.append(None)

df1['phone_scraped_matches_row_ids'] = matches_list

# Count how many rows got matches
matched_count = df1['phone_scraped_matches_row_ids'].notna().sum()
step_time = time.time() - step_start
print(f"âœ“ Phone number matching complete: {matched_count} rows matched")
print(f"  Time: {step_time:.2f} seconds")
```

```python
# STEP 2: Match by ZIP code (OPTIMIZED)
print(f"\n{'='*60}")
print("STEP 2: ZIP Code Matching (OPTIMIZED)")
print(f"{'='*60}")
step_start = time.time()

# Create the new column to store matching row IDs
df1['ZIP_scraped_matches_row_ids'] = None

# Function to clean ZIP codes by ensuring they're strings and removing extra spaces
def clean_zip_code(zip_str):
    if pd.isna(zip_str):
        return ''
    # Convert to string and strip any leading/trailing spaces
    zip_clean = str(zip_str).strip()
    # Take only the first 5 digits for US ZIP codes
    zip_clean = re.sub(r'[^0-9]', '', zip_clean)[:5]
    return zip_clean

# Clean ZIP codes in df1
df1_clean_zips = df1['zip_code'].apply(clean_zip_code)

# Clean ZIP codes in df2
df2_clean_zips = df2['Postal Code'].apply(clean_zip_code) if 'Postal Code' in df2.columns else pd.Series(dtype=str)

# If Postal Code column doesn't exist, alert the user
if 'Postal Code' not in df2.columns:
    print("Warning: 'Postal Code' column not found in df2")

# OPTIMIZATION: Build reverse index - ZIP code to df2 row IDs
print("Building ZIP code index...")
zip_to_df2_rows = {}
for j, zip_code in enumerate(df2_clean_zips):
    if zip_code:  # Only index non-empty ZIP codes
        if zip_code not in zip_to_df2_rows:
            zip_to_df2_rows[zip_code] = []
        zip_to_df2_rows[zip_code].append(j)

# Vectorized matching using the index
matches_list = []
for zip_code in df1_clean_zips:
    if zip_code and zip_code in zip_to_df2_rows:
        matches_list.append(str(zip_to_df2_rows[zip_code]))
    else:
        matches_list.append(None)

df1['ZIP_scraped_matches_row_ids'] = matches_list

# Count how many rows got matches
matched_count = df1['ZIP_scraped_matches_row_ids'].notna().sum()
step_time = time.time() - step_start
print(f"âœ“ ZIP code matching complete: {matched_count} rows matched")
print(f"  Time: {step_time:.2f} seconds")
```

```python
# STEP 3: Match by city name with permissive matching (OPTIMIZED)
print(f"\n{'='*60}")
print("STEP 3: City Name Matching (OPTIMIZED)")
print(f"{'='*60}")
step_start = time.time()

# Create the new column to store matching row IDs
df1['City_scraped_matches_row_ids'] = None

# Function to clean and tokenize city names
def clean_city(city_str):
    if pd.isna(city_str):
        return []
    
    # Convert to lowercase, remove special characters
    city_clean = re.sub(r'[^a-zA-Z0-9\s]', '', str(city_str).lower()).strip()
    
    # Remove ordinal suffixes
    city_clean = clean_ordinal_suffixes(city_clean)
    
    # Split into words and filter using centralized common words filter
    words = city_clean.split()
    common_words = get_comprehensive_common_words_filter()
    
    # Filter out common words and keep meaningful city name parts
    filtered_words = []
    for word in words:
        clean_word = re.sub(r'[^a-z0-9]', '', word)
        if (clean_word not in common_words and 
            len(clean_word) > 1 and 
            not clean_word.isdigit()):
            filtered_words.append(clean_word)
    
    return filtered_words

# Clean and tokenize city names in df1
df1_city_tokens = df1['city'].apply(clean_city)
df1_major_city_tokens = df1['major_city'].apply(clean_city) if 'major_city' in df1.columns else pd.Series([[] for _ in range(len(df1))])

# Clean and tokenize city names in df2
if 'City' in df2.columns:
    df2_city_tokens = df2['City'].apply(clean_city)
else:
    print("Warning: 'City' column not found in df2")
    df2_city_tokens = pd.Series([[] for _ in range(len(df2))])

# OPTIMIZATION: Build reverse index - token to df2 row IDs
print("Building city token index...")
city_token_to_df2_rows = {}
for j, tokens in enumerate(df2_city_tokens):
    for token in tokens:
        if token not in city_token_to_df2_rows:
            city_token_to_df2_rows[token] = []
        city_token_to_df2_rows[token].append(j)

# Vectorized matching using the index
matches_list = []
for i in range(len(df1)):
    # Combine tokens from city and major_city
    all_tokens = set(df1_city_tokens.iloc[i] + df1_major_city_tokens.iloc[i])
    
    # Skip if no tokens to match
    if not all_tokens:
        matches_list.append(None)
        continue
    
    # Use index lookup instead of nested loop
    matches = set()
    for token in all_tokens:
        if token in city_token_to_df2_rows:
            matches.update(city_token_to_df2_rows[token])
    
    # Store the matches
    matches_list.append(str(list(matches)) if matches else None)

df1['City_scraped_matches_row_ids'] = matches_list

# Count how many rows got matches
matched_count = df1['City_scraped_matches_row_ids'].notna().sum()
step_time = time.time() - step_start
print(f"City matching complete: {matched_count} rows matched")
print(f"  Time: {step_time:.2f} seconds")
```

```python
# STEP 4: Match by exit number with permissive matching (OPTIMIZED)
print(f"\n{'='*60}")
print("STEP 4: Exit Number Matching (OPTIMIZED)")
print(f"{'='*60}")
step_start = time.time()

# Create the new column to store matching row IDs
df1['Exit_scraped_matches_row_ids'] = None

# Function to extract exit numbers from text
def extract_exit_numbers(exit_str):
    if pd.isna(exit_str):
        return []
    
    # Convert to string
    exit_str = str(exit_str)
    
    # Extract numeric portions that might represent exit numbers
    exit_numbers = []
    
    # Find all numbers in the string
    number_matches = re.findall(r'\d+', exit_str)
    exit_numbers.extend(number_matches)
    
    # Special handling for compound exits like "160 EB/164 WB"
    if '/' in exit_str:
        parts = exit_str.split('/')
        for part in parts:
            part_numbers = re.findall(r'\d+', part)
            if part_numbers:
                exit_numbers.extend(part_numbers)
    
    # Remove duplicates
    return list(set(exit_numbers))

# Extract exit numbers from all relevant columns in df1
exit_columns = ['Exit_Number', 'Exit_From_Address', 'Exit_From_Label', 'Exit_Number_2', 'Exit_Number_3']
df1_exit_numbers = {}

for col in exit_columns:
    if col in df1.columns:
        df1_exit_numbers[col] = df1[col].apply(extract_exit_numbers)
    else:
        print(f"Warning: '{col}' column not found in df1")

# Extract exit numbers from df2
if 'Exit' in df2.columns:
    df2_exit_numbers = df2['Exit'].apply(extract_exit_numbers)
else:
    print("Warning: 'Exit' column not found in df2")
    df2_exit_numbers = pd.Series([[] for _ in range(len(df2))])

# OPTIMIZATION: Build reverse index - exit number to df2 row IDs
print("Building exit number index...")
exit_number_to_df2_rows = {}
for j, exit_nums in enumerate(df2_exit_numbers):
    for num in exit_nums:
        if num not in exit_number_to_df2_rows:
            exit_number_to_df2_rows[num] = []
        exit_number_to_df2_rows[num].append(j)

# Vectorized matching using the index
matches_list = []
for i in range(len(df1)):
    # Collect all exit numbers from this df1 row
    all_exit_numbers = []
    for col, exit_nums in df1_exit_numbers.items():
        if i < len(exit_nums):  # Ensure index is in range
            all_exit_numbers.extend(exit_nums.iloc[i])
    
    # Skip if no exit numbers to match
    if not all_exit_numbers:
        matches_list.append(None)
        continue
    
    # Use index lookup instead of nested loop
    matches = set()
    for num in all_exit_numbers:
        if num in exit_number_to_df2_rows:
            matches.update(exit_number_to_df2_rows[num])
    
    # Store the matches
    matches_list.append(str(list(matches)) if matches else None)

df1['Exit_scraped_matches_row_ids'] = matches_list

# Count how many rows got matches
matched_count = df1['Exit_scraped_matches_row_ids'].notna().sum()
step_time = time.time() - step_start
print(f"Exit number matching complete: {matched_count} rows matched")
print(f"  Time: {step_time:.2f} seconds")
```

```python
# STEP 5: Match by state (OPTIMIZED)
print(f"\n{'='*60}")
print("STEP 5: State Matching (OPTIMIZED)")
print(f"{'='*60}")
step_start = time.time()

# Create the new column to store matching row IDs
df1['State_scraped_matches_row_ids'] = None

# Function to clean state names/abbreviations
def clean_state(state_str):
    if pd.isna(state_str):
        return ''
    # Convert to uppercase and remove spaces/special characters
    return re.sub(r'[^A-Z]', '', str(state_str).upper())

# Clean states in df1
df1_clean_states = df1['state'].apply(clean_state) if 'state' in df1.columns else pd.Series(dtype=str)

# Clean states in df2
df2_clean_states = df2['State'].apply(clean_state) if 'State' in df2.columns else pd.Series(dtype=str)

# If state column doesn't exist in either dataframe, alert the user
if 'state' not in df1.columns:
    print("Warning: 'state' column not found in df1")
if 'State' not in df2.columns:
    print("Warning: 'State' column not found in df2")

# OPTIMIZATION: Build reverse index - state to df2 row IDs
print("Building state index...")
state_to_df2_rows = {}
for j, state in enumerate(df2_clean_states):
    if state:  # Only index non-empty states
        if state not in state_to_df2_rows:
            state_to_df2_rows[state] = []
        state_to_df2_rows[state].append(j)

# Vectorized matching using the index
matches_list = []
for state in df1_clean_states:
    if state and state in state_to_df2_rows:
        matches_list.append(str(state_to_df2_rows[state]))
    else:
        matches_list.append(None)

df1['State_scraped_matches_row_ids'] = matches_list

# Count how many rows got matches
matched_count = df1['State_scraped_matches_row_ids'].notna().sum()
step_time = time.time() - step_start
print(f"âœ“ State matching complete: {matched_count} rows matched")
print(f"  Time: {step_time:.2f} seconds")
```

```python
# STEP 6: Match by road name with permissive matching (OPTIMIZED)
print(f"\n{'='*60}")
print("STEP 6: Road Name Matching (OPTIMIZED)")
print(f"{'='*60}")
step_start = time.time()

# Create the new column to store matching row IDs
df1['Road_scraped_matches_row_ids'] = None

# Function to extract road identifiers and names
def extract_road_tokens(road_str):
    if pd.isna(road_str):
        return set()
    
    road_str = str(road_str).lower()
    tokens = set()
    
    # Extract basic road name with ordinal suffixes cleaned
    basic_name = re.sub(r'[^a-z0-9\s]', ' ', road_str)
    basic_name = clean_ordinal_suffixes(basic_name)
    
    tokens.update(basic_name.split())
    
    # Extract route numbers (e.g., US 60-70, I-10, NV 604)
    route_numbers = re.findall(r'(\d+)(?:\s*-\s*(\d+))?', road_str)
    for match in route_numbers:
        tokens.add(match[0])  # Add first number
        if match[1]:  # Add second number if it exists (e.g., "60-70" â†’ add both "60" and "70")
            tokens.add(match[1])
    
    # Handle interstate notation
    interstate_match = re.search(r'i\s*[-]?\s*(\d+)', road_str)
    if interstate_match:
        tokens.add(interstate_match.group(1))  # Add just the number
    
    # Handle state routes (e.g., "NV 604" â†’ add "604")
    state_route_match = re.search(r'[a-z]{2}\s*[-]?\s*(\d+)', road_str)
    if state_route_match:
        tokens.add(state_route_match.group(1))  # Add just the number
    
    # Handle compound routes with slashes (e.g., "NV 604/574" â†’ add both "604" and "574")
    if '/' in road_str:
        slash_parts = road_str.split('/')
        for part in slash_parts:
            numbers = re.findall(r'\d+', part)
            tokens.update(numbers)
    
    # Filter using centralized common words filter
    words = road_str.split()
    filtered_words = []
    common_words = get_comprehensive_common_words_filter()
    
    for word in words:
        # Remove ordinal suffixes from individual words
        clean_word = clean_ordinal_suffixes(word)
        clean_word = re.sub(r'[^a-z0-9]', '', clean_word)
        
        if (clean_word not in common_words and 
            not (len(clean_word) == 1 and clean_word.isalpha()) and  # Skip single letters like N, S, E, W
            len(clean_word) > 0):  # Skip empty strings
            filtered_words.append(clean_word)
    
    tokens.update(filtered_words)
    
    # Remove common words using centralized filter
    tokens = tokens - common_words
    
    return tokens

# Get road tokens from df1
road_columns = ['Main_Road', 'Secondary_Road', 'Tertiary_Road']
df1_road_tokens = {}

for col in road_columns:
    if col in df1.columns:
        df1_road_tokens[col] = df1[col].apply(extract_road_tokens)
    else:
        print(f"Warning: '{col}' column not found in df1")

# Get road tokens from df2
df2_road_columns = ['Highway', 'Street Address', 'Mailing Address', 'Road Name']
df2_road_tokens = {}

for col in df2_road_columns:
    if col in df2.columns:
        df2_road_tokens[col] = df2[col].apply(extract_road_tokens)
    else:
        print(f"Warning: '{col}' column not found in df2")

# OPTIMIZATION: Pre-compute all df2 tokens and build reverse index
print("Building road token index...")
df2_all_road_tokens = []
road_token_to_df2_rows = {}

for j in range(len(df2)):
    # Collect all road tokens from this df2 row
    df2_row_tokens = set()
    for col, tokens_series in df2_road_tokens.items():
        if j < len(tokens_series):
            df2_row_tokens.update(tokens_series.iloc[j])
    
    df2_all_road_tokens.append(df2_row_tokens)
    
    # Build reverse index
    for token in df2_row_tokens:
        if token not in road_token_to_df2_rows:
            road_token_to_df2_rows[token] = []
        road_token_to_df2_rows[token].append(j)

# Vectorized matching using the index
matches_list = []
for i in range(len(df1)):
    # Collect all road tokens from this df1 row
    all_road_tokens = set()
    for col, tokens_series in df1_road_tokens.items():
        if i < len(tokens_series):  # Ensure index is in range
            all_road_tokens.update(tokens_series.iloc[i])
    
    # Skip if no road tokens to match
    if not all_road_tokens:
        matches_list.append(None)
        continue
    
    # Use index lookup instead of nested loop
    matches = set()
    for token in all_road_tokens:
        if token in road_token_to_df2_rows:
            matches.update(road_token_to_df2_rows[token])
    
    # Store the matches
    matches_list.append(str(list(matches)) if matches else None)

df1['Road_scraped_matches_row_ids'] = matches_list

# Count how many rows got matches
matched_count = df1['Road_scraped_matches_row_ids'].notna().sum()
step_time = time.time() - step_start
print(f"Road name matching complete: {matched_count} rows matched")
print(f"  Time: {step_time:.2f} seconds")
```

```python
# STEP 7: Match by gas station chain with permissive matching (OPTIMIZED)
print(f"\n{'='*60}")
print("STEP 7: Chain Name Matching (OPTIMIZED)")
print(f"{'='*60}")
step_start = time.time()

# Create the new column to store matching row IDs
df1['Chain_scraped_matches_row_ids'] = None

# Function to standardize chain names for better matching
def standardize_chain(chain_str):
    if pd.isna(chain_str):
        return set()
    
    chain_str = str(chain_str).lower()
    tokens = set()
    
    # Standard mapping of chain variations to canonical forms
    chain_mappings = {
        'sinclair': ['sinclair', 'sinclair oil'],
        'travelcenters': ['ta', 'taexpress', 'talogo'],
        'eleven': ['11', '7-11', 'seven', 'eleven', '7 eleven', '7eleven'],
        "love's": ['love', 'loves', "love's"],
        'speedway': ['speedway', 'speedwaygas'],
        'chevron': ['chevron'],
        'flying j': ['flying j'],
        'tesoro': ['tesoro'],
        'shell': ['shell'],
        'texaco': ['texaco'],
        'exxon': ['exxon'],
        'pilot': ['pilot'],
        'conoco': ['conoco'],
        'shamrock': ['shamrock'],
        'valero': ['valero'],
        'bp': ['bp'],
        'mobil': ['mobil'],
        'circle k': ['circle_k', 'circle k'],
        'citgo': ['citgo'],
        '76': ['76', 'union 76']
    }
    
    # Add the original chain text
    tokens.add(chain_str)
    
    # Add individual words
    words = re.sub(r'[^a-z0-9\s]', ' ', chain_str).split()
    tokens.update(words)
    
    # Add known variations based on mapping
    for canonical, variations in chain_mappings.items():
        for variation in variations:
            if variation in chain_str:
                tokens.add(canonical)
                tokens.update(variations)
    
    # Specific handling for "7-Eleven" / "Eleven" because it appears in different formats
    if any(term in chain_str for term in ['11', 'seven', 'eleven']):
        tokens.update(['11', '7-11', 'seven', 'eleven', '7 eleven', '7eleven'])
    
    return tokens

# Get chain tokens from df1
if 'chain' in df1.columns:
    df1_chain_tokens = df1['chain'].apply(standardize_chain)
else:
    print("Warning: 'chain' column not found in df1")
    df1_chain_tokens = pd.Series([set() for _ in range(len(df1))])

# Get chain tokens from df2
if 'Chain' in df2.columns:
    df2_chain_tokens = df2['Chain'].apply(standardize_chain)
else:
    print("Warning: 'Chain' column not found in df2")
    df2_chain_tokens = pd.Series([set() for _ in range(len(df2))])

# OPTIMIZATION: Build reverse index for chain tokens
print("Building chain token index...")
chain_token_to_df2_rows = {}
for j, tokens in enumerate(df2_chain_tokens):
    for token in tokens:
        if token not in chain_token_to_df2_rows:
            chain_token_to_df2_rows[token] = []
        chain_token_to_df2_rows[token].append(j)

# Vectorized matching using the index
matches_list = []
for i in range(len(df1)):
    df1_tokens = df1_chain_tokens.iloc[i]
    
    # Skip if no tokens to match
    if not df1_tokens:
        matches_list.append(None)
        continue
    
    # Use index lookup instead of nested loop
    matches = set()
    for token in df1_tokens:
        if token in chain_token_to_df2_rows:
            matches.update(chain_token_to_df2_rows[token])
    
    # Store the matches
    matches_list.append(str(list(matches)) if matches else None)

df1['Chain_scraped_matches_row_ids'] = matches_list

# Count how many rows got matches
matched_count = df1['Chain_scraped_matches_row_ids'].notna().sum()
step_time = time.time() - step_start
print(f"Chain matching complete: {matched_count} rows matched")
print(f"  Time: {step_time:.2f} seconds")
```

```python
# STEP 8: Match by label text (OPTIMIZED)
print(f"\n{'='*60}")
print("STEP 8: Label Text Matching (OPTIMIZED)")
print(f"{'='*60}")
step_start = time.time()

# Create the new column to store matching row IDs
df1['Label_scraped_matches_row_ids'] = None

# Function to extract meaningful words from labels
def extract_label_words(label_str):
    if pd.isna(label_str):
        return set()
    
    # Convert to lowercase and remove special characters
    clean_label = re.sub(r'[^a-z0-9\s]', ' ', str(label_str).lower())
    
    # Remove ordinal suffixes using centralized function
    clean_label = clean_ordinal_suffixes(clean_label)
    
    # Split into words
    words = clean_label.split()
    
    # Use centralized common words filter
    common_words = get_comprehensive_common_words_filter()
    
    meaningful_words = set()
    for word in words:
        # Clean and remove ordinal suffixes from individual words
        clean_word = clean_ordinal_suffixes(word)
        clean_word = re.sub(r'[^a-z0-9]', '', clean_word)
        
        # Skip if it's a common word, single digit, or too short
        if (clean_word not in common_words and 
            not (clean_word.isdigit() and len(clean_word) == 1) and 
            len(clean_word) > 1):
            meaningful_words.add(clean_word)
            
            # Handle special case for "7-Eleven" type stores
            if clean_word == '7' and any(eleven_word in words for eleven_word in ['eleven', '11']):
                meaningful_words.add('7-eleven')
            if clean_word == 'loves':
                meaningful_words.add("love's")
    
    return meaningful_words

# Get label tokens from df1
if 'label' in df1.columns:
    df1_label_tokens = df1['label'].apply(extract_label_words)
else:
    print("Warning: 'label' column not found in df1")
    df1_label_tokens = pd.Series([set() for _ in range(len(df1))])

# Get name and chain tokens from df2
df2_name_tokens = set()
df2_chain_tokens_for_label = set()

if 'name' in df2.columns:
    df2_name_tokens = df2['name'].apply(extract_label_words)
else:
    print("Warning: 'name' column not found in df2")
    df2_name_tokens = pd.Series([set() for _ in range(len(df2))])

if 'Chain' in df2.columns:
    df2_chain_tokens_for_label = df2['Chain'].apply(extract_label_words)
else:
    print("Warning: 'Chain' column not found in df2")
    df2_chain_tokens_for_label = pd.Series([set() for _ in range(len(df2))])

# OPTIMIZATION: Pre-compute combined tokens and build reverse index
print("Building label token index...")
df2_combined_label_tokens = []
label_token_to_df2_rows = {}

for j in range(len(df2)):
    # Combine name and chain tokens from df2
    combined_tokens = df2_name_tokens.iloc[j].union(df2_chain_tokens_for_label.iloc[j])
    df2_combined_label_tokens.append(combined_tokens)
    
    # Build reverse index
    for token in combined_tokens:
        if token not in label_token_to_df2_rows:
            label_token_to_df2_rows[token] = []
        label_token_to_df2_rows[token].append(j)

# Vectorized matching using the index
matches_list = []
for i in range(len(df1)):
    df1_tokens = df1_label_tokens.iloc[i]
    
    # Skip if no tokens to match
    if not df1_tokens:
        matches_list.append(None)
        continue
    
    # Use index lookup instead of nested loop
    matches = set()
    for token in df1_tokens:
        if token in label_token_to_df2_rows:
            matches.update(label_token_to_df2_rows[token])
    
    # Store the matches
    matches_list.append(str(list(matches)) if matches else None)

df1['Label_scraped_matches_row_ids'] = matches_list

# Count how many rows got matches
matched_count = df1['Label_scraped_matches_row_ids'].notna().sum()
step_time = time.time() - step_start
print(f"Label text matching complete: {matched_count} rows matched")
print(f"  Time: {step_time:.2f} seconds")
```

```python
# FINAL STEP: Summarize matches and save results
print(f"\n{'='*60}")
print("FINAL STEP: Summarizing and Saving Results")
print(f"{'='*60}")

# Convert string representations of lists to actual lists
def str_to_list(str_val):
    # Handle None and NaN values
    if str_val is None:
        return []
    
    # Check if it's already a list (shouldn't happen, but just in case)
    if isinstance(str_val, list):
        return str_val
    
    # Handle numpy/pandas NA values using try-except
    try:
        # Check if it's a string
        if not isinstance(str_val, str):
            return []
        
        # Empty string check
        if str_val == '' or str_val == 'None':
            return []
        
        # Try to evaluate the string as a literal Python expression
        return eval(str_val)
    except:
        # If that fails, return an empty list
        return []

# Convert all matching columns to actual lists
match_columns = [
    'phone_scraped_matches_row_ids',
    'ZIP_scraped_matches_row_ids',
    'City_scraped_matches_row_ids',
    'Exit_scraped_matches_row_ids',
    'State_scraped_matches_row_ids',
    'Road_scraped_matches_row_ids',
    'Chain_scraped_matches_row_ids',
    'Label_scraped_matches_row_ids'
]

for col in match_columns:
    if col in df1.columns:
        df1[col] = df1[col].apply(str_to_list)

# Create a new column that combines all matches
df1['all_matches'] = df1.apply(
    lambda row: list(set().union(*[row[col] for col in match_columns if col in df1.columns and isinstance(row[col], list)])),
    axis=1
)

# Count matches by type
match_summary = {
    'Total rows in df1': len(df1),
    'Total rows in df2': len(df2)
}

for col in match_columns:
    if col in df1.columns:
        match_summary[f'Rows with {col}'] = sum(df1[col].apply(lambda x: len(x) > 0))

match_summary['Rows with any match'] = sum(df1['all_matches'].apply(lambda x: len(x) > 0))
match_summary['Rows with no matches'] = sum(df1['all_matches'].apply(lambda x: len(x) == 0))

# Calculate percentage of rows matched
match_summary['Percentage matched'] = (match_summary['Rows with any match'] / match_summary['Total rows in df1']) * 100

# Display the summary
for key, value in match_summary.items():
    if 'Percentage' in key:
        print(f"{key}: {value:.2f}%")
    else:
        print(f"{key}: {value}")

# Save the results to Parquet format
print("\nSaving results...")
print("Saving as Parquet...")
temp_output_dir = PROJECT_ROOT / "temp" / "1_derived"
temp_output_dir.mkdir(parents=True, exist_ok=True)
parquet_path = temp_output_dir / "1_merge_station_records.parquet"
csv_path = temp_output_dir / "1_merge_station_records.csv"
try:
    df1.to_parquet(parquet_path, index=False, engine='pyarrow')
    print(f"Saved to {parquet_path}")
    import os
    file_size = os.path.getsize(parquet_path) / (1024 * 1024)  # MB
    print(f"  File size: {file_size:.2f} MB")
except Exception as e:
    print(f"WARNING: Parquet save failed: {e}")
    print("\nTrying alternative: Saving as CSV instead...")
    df1.to_csv(csv_path, index=False)
    print(f"Saved to {csv_path}")
    import os
    file_size = os.path.getsize(csv_path) / (1024 * 1024)  # MB
    print(f"  File size: {file_size:.2f} MB")
    print("\nNOTE: To fix Parquet issue: Restart the kernel and re-run all cells.")

# Calculate total execution time
end_time = time.time()
elapsed_time = end_time - start_time
print(f"\n{'='*60}")
print(f"Total execution time: {elapsed_time:.2f} seconds ({elapsed_time/60:.2f} minutes)")
print(f"Finished at: {time.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"{'='*60}")

# Preview sample rows with matches
print("\nSample rows with matches:")
sample_with_matches = df1[df1['all_matches'].apply(lambda x: len(x) > 0)].head(5)
display(sample_with_matches[['label', 'phone', 'city', 'state', 'chain', 'all_matches']])
```
